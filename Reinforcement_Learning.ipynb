No campo do Machine Learning, o **Aprendizado por Reforço** ainda não é amplamente aplicado em aplicações comerciais hoje em dia, mas é um dos pilares fundamentais dessa área. Além disso, há muitas pesquisas empolgantes que o respaldam e o aprimoram a cada dia.

Assim como muitos outros helicópteros autônomos, ele é equipado com um computador de bordo, GPS, acelerômetros, giroscópios e uma bússola magnética. Isso permite que ele saiba sua posição exata em todos os momentos.

## O Desafio do Controle

Os helicópteros controlados por rádio utilizam **joysticks** como este:

image.png

- A cada **décimo de segundo**, você recebe informações sobre a posição, orientação e velocidade do helicóptero.
- Sua tarefa é decidir como mover os dois **sticks de controle** para manter o helicóptero equilibrado no ar.

Então, como escrever um programa que faça isso automaticamente?

# Como Ensinar um Helicóptero a Voar Sozinho com Aprendizado por Reforço?

A tarefa consiste em, dado o estado do helicóptero (posição, orientação, velocidade, etc.), decidir como mover os **sticks de controle**.

image.png

No **Aprendizado por Reforço**, chamamos esse conjunto de informações de **estado (s)**. O objetivo é encontrar uma **função** que mapeie esse estado para uma **ação (a)**, ou seja, determinar como mover os sticks de controle para manter o helicóptero equilibrado, voando corretamente e sem cair.

## Por que Não Usar Aprendizado Supervisionado?

Uma abordagem inicial poderia ser utilizar **Aprendizado Supervisionado**. No entanto, essa estratégia **não funciona bem** para helicópteros autônomos.

A ideia seria:

- Coletar várias observações de diferentes **estados (s)** do helicóptero.
- Pedir a um piloto especialista para fornecer a **melhor ação (y)** a ser tomada em cada estado.
- Treinar uma rede neural para aprender a mapear **estados (s)** para **ações (a)**, como um problema de classificação ou regressão.

image.png

**Por que essa abordagem falha?**
- Quando o helicóptero está no ar, **há muita ambiguidade** sobre a ação ideal a ser tomada.
- Pequenas variações podem levar a diferentes resultados:  
  - Incliná-lo um pouco para a esquerda ou muito para a esquerda?
  - Aumentar ligeiramente a potência ou muito?
- É extremamente difícil criar um conjunto de dados **preciso** e **completo** contendo os estados e as ações corretas.

Por isso, para tarefas como o controle de helicópteros e outros robôs, o **Aprendizado Supervisionado** não é eficaz. Em vez disso, utilizamos **Aprendizado por Reforço**.

## O Papel da Recompensa no Aprendizado por Reforço

Um dos principais conceitos do **Aprendizado por Reforço** é a **recompensa** ou **função de recompensa**.

Essa função indica ao helicóptero **quando ele está indo bem e quando está indo mal**. Em outras palavras, o sistema de aprendizado recebe **feedback** sobre suas ações:

✅ **Recompensa positiva** → Se o helicóptero se mantém equilibrado e voando corretamente.  
❌ **Recompensa negativa** → Se o helicóptero perde estabilidade, oscila demais ou cai.  

O algoritmo, então, aprende a maximizar a **soma das recompensas** ao longo do tempo, ajustando suas ações para melhorar o desempenho.

# A Função de Recompensa: Treinando um Helicóptero Como se Fosse um Cachorro  

Uma maneira intuitiva de pensar sobre a **função de recompensa** no Aprendizado por Reforço é compará-la ao treinamento de um cachorro.  

Quando eu era criança, minha família tinha um cachorro, e minha tarefa era treiná-lo para se comportar bem. Mas como ensinar um filhote a se comportar corretamente?  

- **Não dá para demonstrar tudo diretamente ao cachorro.**  
- Em vez disso, deixamos ele agir livremente e fornecemos feedback:  
  - **Se ele faz algo bom** → Dizemos "bom cachorro!" 🐶✅  
  - **Se ele faz algo ruim** → Dizemos "cachorro mau!" ❌  

Com o tempo, o cachorro aprende a **fazer mais das coisas boas** e **menos das ruins**.  

## Como Isso se Relaciona ao Aprendizado por Reforço?  

O treinamento com um **algoritmo de Aprendizado por Reforço** funciona da mesma forma:  

- **Se o helicóptero está voando bem** → Dizemos "bom helicóptero!" 🚁✅  
- **Se ele começa a oscilar ou cai** → Dizemos "helicóptero ruim!" ❌  

O **algoritmo de Aprendizado por Reforço** então ajusta suas ações para **maximizar as recompensas positivas** e **minimizar as penalizações negativas**.  

## Por Que o Aprendizado por Reforço é Tão Poderoso?

O Aprendizado por Reforço se destaca porque **não precisamos dizer ao sistema exatamente como realizar uma tarefa**. Em vez disso, **apenas dizemos o que ele deve alcançar**.  

Em termos práticos, para controlar um helicóptero:  

- **Se ele está voando bem**, pode receber **+1 ponto de recompensa por segundo**.  
- **Se começa a voar de forma instável**, pode receber uma **pequena penalização negativa**.  
- **Se ele cair**, pode receber uma **grande penalização**, como **-1000 pontos**.  

Dessa forma, o helicóptero **aprende sozinho** a **manter o voo estável** e **evitar quedas**, pois isso maximiza sua recompensa ao longo do tempo.

# Mars rover example

## **Exemplo Simplificado de Aprendizado por Reforço: Rover em Marte 🚀**  

Para finalizar a formalização do **Aprendizado por Reforço**, em vez de analisarmos algo complexo como um helicóptero ou um cão robótico, podemos usar um exemplo mais simples, inspirado no **rover de Marte**.  

Este exemplo é baseado no trabalho da professora **Emma Branskill** de Stanford e de **Jagriti Agrawal**, que ajudou a desenvolver este modelo e também escreveu códigos que atualmente controlam o **rover real em Marte**.  

## **Estados e Recompensas no Ambiente**  

O rover pode estar em **um dos seis estados possíveis**, representados pelos seis quadrados na imagem.  

image.png

📍 O rover começa no **estado 4** (conforme indicado na imagem).  

O objetivo do rover é realizar **missões científicas**, como:  
- Coletar amostras de solo usando um **espectrômetro, broca ou radar**.  
- Tirar fotos de áreas interessantes para os cientistas na Terra.  

image.png

Os estados **1 e 6** são locais de interesse científico:  
- **Estado 1** contém uma superfície muito valiosa para análise.  
  - ✅ **Recompensa de +100** (valor mais alto).  
- **Estado 6** também tem uma superfície interessante, mas não tanto quanto o estado 1.  
  - ✅ **Recompensa de +40**.  
- Os estados intermediários **(2, 3, 4 e 5)** não são tão cientificamente úteis.  
  - ❌ **Recompensa de 0**.  

🔹 **Estados 1 e 6 são estados terminais**, o que significa que, ao chegar a um deles, a missão do rover para aquele dia termina.

## **Ações Disponíveis para o Rover**  

image.png

A cada passo, o rover pode escolher entre **duas ações**:  
1. **Ir para a esquerda** ⬅️  
2. **Ir para a direita** ➡️  

A questão é: **qual a melhor estratégia para maximizar as recompensas?**  

## **Caminhos Possíveis e Recompensas**  

### 🔵 **Caminho indo para a esquerda (em direção ao estado 1)**  

1️⃣ Começando no **estado 4** → **Recompensa = 0**  
2️⃣ Move-se para o **estado 3** → **Recompensa = 0**  
3️⃣ Move-se para o **estado 2** → **Recompensa = 0**  
4️⃣ Chega ao **estado 1** → **Recompensa = +100** ✅ (fim da missão)  

### 🟠 **Caminho indo para a direita (em direção ao estado 6)**  

1️⃣ Começando no **estado 4** → **Recompensa = 0**  
2️⃣ Move-se para o **estado 5** → **Recompensa = 0**  
3️⃣ Move-se para o **estado 6** → **Recompensa = +40** ✅ (fim da missão)  

🔹 **Conclusão:** O caminho para **o estado 1 maximiza a recompensa (100 vs. 40)**, então seria preferível para o rover escolher esse trajeto.  

## **Decisões Ineficientes**  

O rover pode decidir **mudar de direção durante a navegação**, mas isso pode desperdiçar tempo e recursos.  

Exemplo:  
1️⃣ Vai da **posição 4 → 5** (direita, recompensa = 0).  
2️⃣ Depois volta para **4 → 3 → 2 → 1** (esquerda, recompensa final = 100).  

🛑 **Essa estratégia é ineficiente**, pois o rover gastou passos extras para atingir um estado que já poderia ter alcançado diretamente.  

## **Elementos Fundamentais do Aprendizado por Reforço**  

Cada decisão do rover segue uma estrutura básica no **Aprendizado por Reforço**:  

- **Estado (S):** Onde o rover está no momento.  
- **Ação (A):** Escolher mover-se para a esquerda ou para a direita.  
- **Recompensa (R):** O valor associado ao estado onde ele se encontra.  
- **Novo Estado (S’):** O estado em que o rover se encontra após tomar uma ação.  

🔹 Por exemplo, no primeiro passo:  
- O rover está no **estado 4**.  
- Toma a ação de **ir para a esquerda**.  
- Recebe uma **recompensa de 0**.  
- Agora está no **estado 3**.  

Essa estrutura (estado, ação, recompensa, próximo estado) é **a base de todos os algoritmos de Aprendizado por Reforço**.  

# The Return in reinforcement learning

Como podemos saber se um conjunto específico de recompensas é **melhor ou pior** do que outro conjunto de recompensas?  

No Aprendizado por Reforço, o conceito de **retorno** nos permite capturar essa ideia.  

Uma analogia útil para entender esse conceito é a seguinte:  

Imagine que você tem uma nota de **cinco dólares** aos seus pés, que pode pegar imediatamente.  
Ou, há uma nota de **dez dólares do outro lado da cidade**, mas seria necessário andar **meia hora** para pegá-la.  

Qual das duas opções você escolheria?  

A nota de dez dólares **vale mais** do que a de cinco, mas se for preciso caminhar por **meia hora** para pegá-la, pode ser mais conveniente simplesmente pegar os cinco dólares disponíveis.  

O **conceito de retorno** captura exatamente essa ideia:  
✔️ Recompensas que podem ser obtidas **mais rapidamente** podem ser mais atraentes.  
✔️ Recompensas **distantes no futuro** podem ter **menos valor** do que recompensas imediatas.  

Agora, vamos ver como isso funciona no exemplo do **rover em Marte**.  

## **Exemplo do Rover em Marte e Cálculo do Retorno**  

image.png

O **rover pode estar em um de seis estados diferentes**, conforme mostrado na imagem.  

📍 O **estado inicial é o estado 4**.  

O objetivo do rover é completar **missões científicas**, como:  
- Coletar amostras de rochas utilizando **espectrômetro, broca ou radar**.  
- Tirar **fotos científicas** para os pesquisadores na Terra.  

As recompensas são distribuídas da seguinte forma:  
- **Estado 1** possui uma superfície **muito valiosa** para estudo.  
  - ✅ **Recompensa = 100**.  
- **Estado 6** também tem uma superfície **interessante, mas não tanto quanto a do estado 1**.  
  - ✅ **Recompensa = 40**.  
- Os estados intermediários **(2, 3, 4 e 5)** **não são tão relevantes** cientificamente.  
  - ❌ **Recompensa = 0**.  

🔹 Os estados **1 e 6 são estados terminais**. Isso significa que, ao chegar a um desses estados, a missão do rover **termina para aquele dia**.  

Agora, vamos calcular o **retorno** que o rover pode obter ao seguir diferentes caminhos.  

## **Cálculo do Retorno com Fator de Desconto (γ)**  

🔹 Se o rover começar no **estado 4** e seguir **para a esquerda**:  
1️⃣ **Estado 4 → Estado 3 → Estado 2 → Estado 1 (terminal, recompensa = 100)**  
2️⃣ Todas as recompensas anteriores **são zero**.  
3️⃣ O retorno é a **soma dessas recompensas**, mas **ponderado pelo fator de desconto (γ)**.  

### **O que é o Fator de Desconto γ?**  
O **fator de desconto** é um número **ligeiramente menor do que 1**, que reduz o valor das recompensas conforme elas estão **mais distantes no futuro**.  

Vamos assumir **γ = 0.9**.  

A fórmula geral do retorno é:  

$$
G = R_1 + \gamma R_2 + \gamma^2 R_3 + \gamma^3 R_4 + \dots
$$

Aplicando **γ = 0.9** no nosso exemplo:  

$$
G = 0 + (0.9 \times 0) + (0.9^2 \times 0) + (0.9^3 \times 100)
$$

$$
G = 0.729 \times 100 = 72.9
$$

Portanto, **com γ = 0.9, o retorno esperado é 72.9**.  

---

## **Interpretando o Fator de Desconto (γ)**  

O **fator de desconto** tem um impacto direto na forma como o algoritmo de Aprendizado por Reforço **valoriza recompensas futuras**:  

✔️ **Se γ for alto (0.99 ou 0.999):**  
- O agente se torna **mais paciente**, pois valoriza **recompensas de longo prazo**.  

✔️ **Se γ for baixo (0.5, por exemplo):**  
- O agente se torna **mais impaciente**, pois prioriza **recompensas imediatas** e dá pouco valor às futuras.  

Para ilustrar isso, vamos calcular o retorno usando **γ = 0.5**.  

## **Cálculo do Retorno com γ = 0.5**  

Se γ for **0.5**, o impacto das recompensas futuras **diminui drasticamente**.  

Aplicando a fórmula:  

$$
G = 0 + (0.5 \times 0) + (0.5^2 \times 0) + (0.5^3 \times 100)
$$

$$
G = 12.5
$$

🔹 **Com um γ menor, o retorno se torna muito menor (12.5 em vez de 72.9)**, pois as recompensas futuras perdem muito valor rapidamente.  

# **Interpretação do Fator de Desconto em Aplicações Financeiras e Retorno no Aprendizado por Reforço 🚀**  

Em **aplicações financeiras**, o **fator de desconto** tem uma interpretação muito intuitiva: ele pode ser visto como a **taxa de juros** ou o **valor do dinheiro no tempo**.  

Por exemplo:  
✔️ Se você pode ter **um dólar hoje**, esse dólar pode valer **mais do que um dólar no futuro**.  
✔️ Isso ocorre porque um dólar hoje pode ser **investido no banco**, gerando **juros** e aumentando de valor ao longo do tempo.  

📌 **No aprendizado por reforço, o fator de desconto desempenha um papel semelhante**: ele reduz o valor das recompensas futuras, tornando-as menos valiosas do que recompensas obtidas mais rapidamente.  

Agora, vamos analisar **como o retorno varia dependendo das ações tomadas pelo agente**.  

## **O Retorno Depende das Ações que Você Toma**  

Como o **retorno** (soma das recompensas futuras descontadas) depende diretamente das **ações do agente**, diferentes estratégias levam a diferentes valores de retorno.  

Vamos usar nosso **exemplo do rover em Marte** novamente, assumindo um fator de desconto **γ = 0.5**.  

## **Cálculo do Retorno Seguindo Sempre para a Esquerda ⬅️**  

Se o rover **sempre for para a esquerda**, os retornos para cada estado serão:  

| Estado  | Recompensa | Retorno |
|---------|------------|------------|
| 1️⃣ | **100** | **100** |
| 2️⃣ | 0 | **50** |
| 3️⃣ | 0 | **25** |
| 4️⃣ | 0 | **12.5** |
| 5️⃣ | 0 | **6.25** |
| 6️⃣ (Terminal) | **40** | **40** |

📌 **Explicação**:  
✔️ Se o agente **começa no estado 1**, ele recebe **100 imediatamente**, sem desconto.  
✔️ Se **começa no estado 2**, ele chega ao estado 1 **um passo depois**, então a recompensa de 100 é multiplicada por **0.5** (γ = 0.5), resultando em **50**.  
✔️ Se **começa no estado 3**, a recompensa de 100 é multiplicada por **0.5² = 0.25**, resultando em **25**.  
✔️ E assim por diante.  

## **Cálculo do Retorno Seguindo Sempre para a Direita ➡️**  

Agora, suponha que o rover **sempre vá para a direita**. Os retornos serão diferentes:  

| Estado  | Recompensa | Retorno |
|---------|------------|------------|
| 1️⃣ | **100** | **100** |
| 2️⃣ | 0 | **2.5** |
| 3️⃣ | 0 | **5** |
| 4️⃣ | 0 | **10** |
| 5️⃣ | 0 | **20** |
| 6️⃣ (Terminal) | **40** | **40** |

📌 **Explicação**:  
✔️ Se o agente **começa no estado 4 e sempre vai para a direita**, ele chega ao estado terminal **6** e recebe **40** como recompensa final.  
✔️ O retorno é então calculado como:  

$$
G = 0 + (0.5 \times 0) + (0.5^2 \times 40) = 10
$$

✔️ Se começa no estado 5, ele chega ao estado 6 **um passo mais cedo**, então o retorno é maior:  

$$
G = 0 + (0.5 \times 40) = 20
$$

## **Comparando Estratégias**  

🔹 **Seguir sempre para a esquerda** maximiza a recompensa final na maioria dos estados, já que o **estado 1 (recompensa 100) é mais valioso** do que o estado 6 (recompensa 40).  
🔹 **Seguir sempre para a direita** leva a retornos menores na maioria dos estados, porque o agente **demora mais para obter recompensas significativas**.  


## **Uma Estratégia Melhor: Mistura de Ações**  

image.png

Em vez de **sempre ir para a esquerda ou para a direita**, podemos **ajustar o comportamento do rover dependendo do estado**.  

**Nova Estratégia:**  
- Se o rover está nos estados **1, 2, 3 ou 4**, ele deve ir para **a esquerda**.  
- Se o rover está no **estado 5**, ele está **próximo do estado 6 (recompensa 40)**, então ele deve ir **para a direita**.  

Os retornos resultantes dessa estratégia mista são:  

| Estado  | Recompensa | Retorno |
|---------|------------|------------|
| 1️⃣ | **100** | **100** |
| 2️⃣ | 0 | **50** |
| 3️⃣ | 0 | **25** |
| 4️⃣ | 0 | **12.5** |
| 5️⃣ | 0 | **20** |
| 6️⃣ (Terminal) | **40** | **40** |

📌 **Por que isso funciona?**  
✔️ O rover segue **para o melhor estado possível dependendo da situação**.  
✔️ Se ele está **longe do estado 6**, ele tenta chegar ao estado 1.  
✔️ Se já está **perto do estado 6**, ele prefere **pegar a recompensa de 40 rapidamente**.  


# Making decisions: Policies in reinforcement learning

Nos algoritmos de **Aprendizado por Reforço (Reinforcement Learning - RL)**, **como um agente decide quais ações tomar?**  

Vamos aprender sobre **o conceito de política (π)** no aprendizado por reforço.  

## **Diferentes Estratégias para Escolher Ações**  

Vimos que há **várias maneiras** de escolher ações dentro de um problema de aprendizado por reforço. Algumas possibilidades incluem:  

image.png

✔️ **Ir sempre para a recompensa mais próxima** → Se a recompensa mais próxima está à esquerda, vá para a esquerda. Se está à direita, vá para a direita.  

image.png

✔️ **Ir sempre para a maior recompensa** → Independentemente da proximidade, escolha a ação que leva à maior recompensa.  

image.png

✔️ **Ir sempre para a menor recompensa** → Embora isso **não seja uma estratégia ideal**, também é uma possibilidade.  

image.png

✔️ **Ir para a esquerda, a menos que esteja a um passo da menor recompensa** → Neste caso, o agente segue uma lógica híbrida.  

## **O Conceito de Política (π) no Aprendizado por Reforço**  

📌 **O que é uma política (π)?**  

No aprendizado por reforço, nosso **objetivo** é definir uma **função chamada política (π)**.  

A política **recebe um estado (s) como entrada** e **mapeia esse estado para uma ação (a)** que o agente deve tomar.  

📌 **Formalmente**:  

$$
\pi(s) = a
$$

🔹 **Exemplo de política (π) da última imagem:**  
- Se **s = 2**, então π(2) = **"←" (esquerda)**.  
- Se **s = 3**, então π(3) = **"←" (esquerda)**.  
- Se **s = 4**, então π(4) = **"←" (esquerda)**.  
- Se **s = 5**, então π(5) = **"→" (direita)**.  

💡 **Ou seja, a política nos diz qual ação o agente deve tomar em cada estado.**  

## **Objetivo do Aprendizado por Reforço**  

O **objetivo do Aprendizado por Reforço** é encontrar **a melhor política (π)** que maximize o **retorno esperado** do agente ao longo do tempo.  

📌 **Isso significa que queremos aprender uma política que nos diga qual ação tomar em cada estado para obter a maior recompensa total possível.**  

💡 Embora o nome "política" (policy) seja o termo padrão na área de aprendizado por reforço, talvez **"controlador"** fosse uma terminologia mais intuitiva. No entanto, "política" é o termo amplamente utilizado na literatura acadêmica.  

# Review of key concepts

Desenvolvemos um formalismo de aprendizagem por reforço usando o exemplo do rover em Marte com seis estados. Vamos fazer uma revisão rápida dos conceitos-chave e também ver como esse conjunto de conceitos pode ser aplicado em outras situações.

Alguns dos conceitos que discutimos incluem os estados de um problema de aprendizagem por reforço, o conjunto de ações, as recompensas, um fator de desconto, como as recompensas e o fator de desconto são usados para calcular o retorno e, finalmente, uma política cuja função é ajudar a escolher as ações de forma a maximizar esse retorno.

No exemplo do rover em Marte, tivemos seis estados numerados de 1 a 6, e as ações disponíveis eram mover-se para a esquerda ou para a direita. As recompensas eram definidas como 100 para o estado mais à esquerda, 40 para o estado mais à direita e zero para os estados intermediários, utilizando um fator de desconto de 0,5. O retorno era calculado por meio de uma fórmula que combinava as recompensas com o fator de desconto, e diferentes políticas (π) poderiam ser adotadas, dependendo do estado em que o agente se encontrava. Esse mesmo formalismo — estados, ações, recompensas, fator de desconto, retorno e política — pode ser aplicado a muitas outras aplicações.

Por exemplo, considere o problema de encontrar um helicóptero autônomo: o estado seria o conjunto de possíveis posições, orientações, velocidades, etc., do helicóptero; as ações corresponderiam às diferentes formas de mover o manche de controle; e as recompensas poderiam ser definidas como +1 se o helicóptero estiver voando bem, ou -1.000 se ele cair de forma grave ou colidir, com um fator de desconto possivelmente definido como 0,99. O retorno seria calculado utilizando a mesma fórmula, e o papel de um algoritmo de aprendizagem por reforço seria encontrar uma política π que, ao receber como entrada a posição do helicóptero, indicasse qual ação tomar, ou seja, como mover os controles.

Outro exemplo é o uso de aprendizagem por reforço para aprender a jogar xadrez, onde o estado do problema seria a posição de todas as peças no tabuleiro (simplificada para este exemplo, mesmo sabendo que outros detalhes também são importantes), as ações seriam todos os movimentos legais possíveis e as recompensas poderiam ser definidas como +1 para uma vitória, -1 para uma derrota e 0 para um empate.

Para o xadrez, normalmente utiliza-se um fator de desconto muito próximo de 1 (por exemplo, 0,99, 0,995 ou 0,999), e o retorno é calculado da mesma forma que nos outros casos, com o objetivo de, a partir de uma dada posição no tabuleiro, escolher a melhor ação usando uma política π. Esse formalismo de aprendizagem por reforço é conhecido como Processo de Decisão de Markov (MDP). Embora o termo possa parecer técnico e complicado, ele se refere exatamente à estrutura discutida: “Markov” indica que o futuro depende apenas do estado atual, e não de como se chegou a esse estado, ou seja, o futuro depende somente de onde você está agora.

Outra forma de compreender esse formalismo é imaginar um robô ou outro agente que desejamos controlar, o qual pode escolher ações (a) e, com base nessas ações, algo acontece no ambiente (como uma mudança de posição ou a execução de uma missão científica). A ação escolhida é definida por uma política π e, em resposta ao que ocorre no mundo, o agente observa o novo estado e recebe as recompensas correspondentes. Muitos autores utilizam diagramas para representar visualmente esse formalismo, mas tais diagramas servem apenas para ilustrar os conceitos que discutimos. Agora, você tem uma compreensão clara de como funciona um problema de aprendizagem por reforço.

# State-action value function definition

## 📌 Definição
Quando começarmos a desenvolver o aprendizado por reforço, veremos que há uma quantidade fundamental que os algoritmos de aprendizado por reforço tentarão calcular: a **função de valor de ação do estado**.

Essa função é representada pela letra maiúscula **Q** e depende do **estado** $s$ em que você pode estar e da **ação** $a$ que você pode escolher nesse estado.

A função $ Q(s, a) $ retorna um número que representa o **retorno** esperado se:
- Você **começar no estado $s$**.
- Tomar a **ação $a$ uma única vez**.
- Depois disso, **se comportar de maneira ótima**.

Ou seja, após tomar a ação $a$, você executará as ações que resultarão no **maior retorno possível**.

## 🤔 Reflexão sobre a definição
Pode parecer estranho definir $Q(s,a)$ dessa forma, pois surge a pergunta: **como sabemos qual é o comportamento ótimo?**  
Se já soubéssemos qual é a política ótima, não precisaríamos calcular $Q(s,a)$, já que já teríamos a política ótima.

Essa definição pode parecer um pouco **circular**, mas não se preocupe! Quando analisarmos os **algoritmos de aprendizado por reforço**, veremos como calcular a função $Q$ **antes mesmo de descobrir a política ótima**.

## 📊 **Exemplo**

image.png

Consideremos um exemplo de um **rover em Marte** onde o **fator de desconto $ \gamma = 0.5 $**.  
Uma política eficiente seria:
- **Ir para a esquerda** nos estados **2, 3 e 4**.
- **Ir para a direita** no estado **5**.

Neste caso, essa estratégia se mostra **ótima**.  
Assim, $Q(s, a)$ será o **retorno total** se você:
- Começar do estado $s$.
- Tomar a ação $a$.
- Em seguida, **seguir a política ótima**.

## Analisando a Função Q do Estado

Vamos analisar a função $ Q(s, a) $ de um estado. Suponha que tomemos a ação de ir para a **direita** a partir do estado 2.

image.png

- Se você estiver no estado 2 e for para a direita, chegará ao estado 3.
- A partir do estado 3, ao se comportar **otimamente**, você irá para a esquerda e voltará para o estado 2.
- Eventualmente, você alcançará o estado terminal e receberá uma recompensa de **100**.

### Recompensas Recebidas
- Estado 2 → **0**
- Estado 3 → **0**
- Estado 2 (de volta) → **0**
- Estado terminal → **100**

O retorno pode ser calculado como:

$$
R = 0 + 0.5 \times 100 + 0.5^2 \times 100 + 0.5^3 \times 100
$$

O que resulta em:

$$
R = 12.5
$$

Assim, temos:

$$
Q(2, \text{direita}) = 12.5
$$

É importante notar que esse valor **não indica** se ir para a direita é uma boa ideia ou não. Ele apenas reporta **o retorno esperado**, assumindo que a ação foi tomada e o comportamento posterior foi ótimo.

## Exemplo: Ir para a Esquerda a partir do Estado 2



Agora, considere a ação de ir **para a esquerda** no estado 2:

- Estado 2 → **0**
- Estado terminal → **100**

O retorno é:

$$
R = 0 + 0.5 \times 100 = 50
$$

Portanto:

$$
Q(2, \text{esquerda}) = 50
$$

No diagrama, podemos representar esses valores como:

- **12.5** à direita, representando $ Q(2, \text{direita}) $
- **50** à esquerda, representando $ Q(2, \text{esquerda}) $

## Exemplo: Ir para a Esquerda a partir do Estado 4

Agora, suponha que estamos no estado 4 e decidimos ir **para a esquerda**:

- Estado 4 → **0**
- Estado intermediário → **0**
- Outro estado intermediário → **0**
- Estado terminal → **100**

O retorno pode ser calculado como:

$$
R = 0 + 0.5 \times 100 + 0.5^2 \times 100 + 0.5^3 \times 100
$$

O que resulta em:

$$
R = 12.5
$$

Portanto:

$$
Q(4, \text{esquerda}) = 12.5
$$


## Valores da Função $ Q(s, a) $

image.png

Se realizarmos este cálculo para todos os estados e todas as ações possíveis, obteremos os valores da **função Q** para cada par $(s, a)$.

No **estado terminal**, a escolha da ação não importa, pois a recompensa final já foi atingida, sendo **100 ou 40**, dependendo do cenário.

A função $ Q(s, a) $ para todos os estados (de 1 a 6) e para as ações (esquerda e direita) pode ser representada de forma organizada.


## O Que é a Função Q?

A função $ Q(s, a) $, também chamada de **função de valor de estado-ação**, responde à pergunta:

> "Qual é o retorno esperado se eu tomar a ação $ a $ no estado $ s $ e depois me comportar de forma ótima?"

Os termos **"função Q"** e **"função de valor de estado-ação"** são frequentemente usados de forma **intercambiável**.


## A Função Q e a Escolha de Ações

Uma vez que podemos calcular a **função Q**, isso nos dá um método para escolher **ações** de forma eficiente.

## Política e Retorno

image.png

A partir dos valores $ Q(s,a) $ analisados anteriormente, podemos observar um padrão interessante ao comparar os diferentes estados:

- **Estado 2**: A ação de ir para a **esquerda** resulta em um valor $ Q(s,a) = 50 $, que é o maior retorno possível a partir desse estado.
- **Estado 3**: A ação de ir para a **esquerda** também resulta no maior retorno.
- **Estado 4**: A ação de ir para a **esquerda** resulta no maior retorno possível.
- **Estado 5**: Neste caso, a **melhor ação** é ir para a **direita**, que dá um retorno de **20**.

Isso nos leva a uma conclusão importante:

> O **melhor retorno possível** de um estado $ s $ é simplesmente o maior valor de $ Q(s,a) $ considerando todas as ações possíveis $ a $.

## Exemplo: Escolhendo a Melhor Ação no Estado 4

No **estado 4**, temos dois valores de $ Q(s,a) $:

- **Q(4, esquerda) = 12.5**
- **Q(4, direita) = 10**

O maior valor entre esses dois é **12.5**, ou seja, **o melhor retorno possível do estado 4 é 12.5**.

Assim, se quisermos que um **robô explorador (Mars Rover)** maximize seu retorno, ele deve escolher a ação $ a $ que leva ao maior valor de $ Q(s,a) $. Nesse caso, a ação **"esquerda"** é a melhor escolha.


## A Importância da Função $ Q(s,a) $

Isso sugere um **princípio fundamental** no aprendizado por reforço:

- Se conseguirmos calcular $ Q(s,a) $ **para todos os estados e ações**, então ao estar em um estado **$ s $**, basta:
  1. Olhar para os diferentes valores de $ Q(s,a) $.
  2. Escolher a ação $ a $ que **maximiza** $ Q(s,a) $.

Isso significa que a **política ótima $ \pi(s) $** pode ser encontrada simplesmente escolhendo a **ação que maximiza** $ Q(s,a) $.

Em resumo:

$$
\pi(s) = \arg\max_{a} Q(s,a)
$$

## Intuição por Trás Dessa Abordagem

A lógica por trás desse método é simples:

- **$ Q(s,a) $** representa o **retorno esperado** se começarmos no estado $ s $, tomarmos a ação $ a $ e, depois disso, **seguirmos uma política ótima**.
- Para obter o **maior retorno possível**, a melhor estratégia é **escolher a ação $ a $ que maximiza $ Q(s,a) $**.

Ou seja:

> **Se soubermos calcular $ Q(s,a) $, escolher a melhor ação se torna um problema trivial: basta pegar a ação que dá o maior valor de $ Q(s,a) $.**


## O Que é $ Q^*(s,a) $?

Embora não seja essencial para este curso, vale mencionar que em muitos textos de aprendizado por reforço, a função $ Q(s,a) $ pode aparecer como **$ Q^*(s,a) $**. Esse termo se refere à **função Q ótima**, que segue exatamente a mesma definição que usamos.

Assim, se encontrar referências a **$ Q^*(s,a) $** ou à **função Q ótima**, saiba que elas significam **o mesmo que a função $ Q(s,a) $ definida aqui**.


# LAB: State-action value function example

````python
# Utils:

import numpy as np
import matplotlib.pyplot as plt

def generate_rewards(num_states, each_step_reward, terminal_left_reward, terminal_right_reward):

    rewards = [each_step_reward] * num_states
    rewards[0] = terminal_left_reward
    rewards[-1] = terminal_right_reward

    return rewards

def generate_transition_prob(num_states, num_actions, misstep_prob = 0):
    # 0 is left, 1 is right

    p = np.zeros((num_states, num_actions, num_states))

    for i in range(num_states):
        if i != 0:
            p[i, 0, i-1] = 1 - misstep_prob
            p[i, 1, i-1] = misstep_prob

        if i != num_states - 1:
            p[i, 1, i+1] = 1  - misstep_prob
            p[i, 0, i+1] = misstep_prob

    # Terminal States
    p[0] = np.zeros((num_actions, num_states))
    p[-1] = np.zeros((num_actions, num_states))

    return p

def calculate_Q_value(num_states, rewards, transition_prob, gamma, V_states, state, action):
    q_sa = rewards[state] + gamma * sum([transition_prob[state, action, sp] * V_states[sp] for sp in range(num_states)])
    return q_sa

def evaluate_policy(num_states, rewards, transition_prob, gamma, policy):
    max_policy_eval = 10000
    threshold = 1e-10

    V = np.zeros(num_states)

    for i in range(max_policy_eval):
        delta = 0
        for s in range(num_states):
            v = V[s]
            V[s] = calculate_Q_value(num_states, rewards, transition_prob, gamma, V, s, policy[s])
            delta = max(delta, abs(v - V[s]))

        if delta < threshold:
            break

    return V

def improve_policy(num_states, num_actions, rewards, transition_prob, gamma, V, policy):
    policy_stable = True

    for s in range(num_states):
        q_best = V[s]
        for a in range(num_actions):
            q_sa = calculate_Q_value(num_states, rewards, transition_prob, gamma, V, s, a)
            if q_sa > q_best and policy[s] != a:
                policy[s] = a
                q_best = q_sa
                policy_stable = False

    return policy, policy_stable


def get_optimal_policy(num_states, num_actions, rewards, transition_prob, gamma):
    optimal_policy = np.zeros(num_states, dtype=int)
    max_policy_iter = 10000

    for i in range(max_policy_iter):
        policy_stable = True

        V = evaluate_policy(num_states, rewards, transition_prob, gamma, optimal_policy)
        optimal_policy, policy_stable = improve_policy(num_states, num_actions, rewards, transition_prob, gamma, V, optimal_policy)

        if policy_stable:
            break

    return optimal_policy, V

def calculate_Q_values(num_states, rewards, transition_prob, gamma, optimal_policy):
    # Left and then optimal policy
    q_left_star = np.zeros(num_states)

    # Right and optimal policy
    q_right_star = np.zeros(num_states)

    V_star =  evaluate_policy(num_states, rewards, transition_prob, gamma, optimal_policy)

    for s in range(num_states):
        q_left_star[s] = calculate_Q_value(num_states, rewards, transition_prob, gamma, V_star, s, 0)
        q_right_star[s] = calculate_Q_value(num_states, rewards, transition_prob, gamma, V_star, s, 1)

    return q_left_star, q_right_star


def plot_optimal_policy_return(num_states, optimal_policy, rewards, V):
    actions = [r"$\leftarrow$" if a == 0 else r"$\rightarrow$" for a in optimal_policy]
    actions[0] = ""
    actions[-1] = ""

    fig, ax = plt.subplots(figsize=(2*num_states,2))

    for i in range(num_states):
        ax.text(i+0.5, 0.5, actions[i], fontsize=32, ha="center", va="center", color="orange")
        ax.text(i+0.5, 0.25, rewards[i], fontsize=16, ha="center", va="center", color="black")
        ax.text(i+0.5, 0.75, round(V[i],2), fontsize=16, ha="center", va="center", color="firebrick")
        ax.axvline(i, color="black")
    ax.set_xlim([0, num_states])
    ax.set_ylim([0, 1])

    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.tick_params(axis='both', which='both', length=0)
    ax.set_title("Optimal policy",fontsize = 16)

def plot_q_values(num_states, q_left_star, q_right_star, rewards):
    fig, ax = plt.subplots(figsize=(3*num_states,2))

    for i in range(num_states):
        ax.text(i+0.2, 0.6, round(q_left_star[i],2), fontsize=16, ha="center", va="center", color="firebrick")
        ax.text(i+0.8, 0.6, round(q_right_star[i],2), fontsize=16, ha="center", va="center", color="firebrick")

        ax.text(i+0.5, 0.25, rewards[i], fontsize=20, ha="center", va="center", color="black")
        ax.axvline(i, color="black")
    ax.set_xlim([0, num_states])
    ax.set_ylim([0, 1])

    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.tick_params(axis='both', which='both', length=0)
    ax.set_title("Q(s,a)",fontsize = 16)

def generate_visualization(terminal_left_reward, terminal_right_reward, each_step_reward, gamma, misstep_prob):
    num_states = 6
    num_actions = 2

    rewards = generate_rewards(num_states, each_step_reward, terminal_left_reward, terminal_right_reward)
    transition_prob = generate_transition_prob(num_states, num_actions, misstep_prob)

    optimal_policy, V = get_optimal_policy(num_states, num_actions, rewards, transition_prob, gamma)
    q_left_star, q_right_star = calculate_Q_values(num_states, rewards, transition_prob, gamma, optimal_policy)

    plot_optimal_policy_return(num_states, optimal_policy, rewards, V)
    plot_q_values(num_states, q_left_star, q_right_star, rewards)
````

# Bellman Equation

Se conseguimos calcular a **função de valor de estado-ação** $ Q(s, a) $, isso nos permite escolher **boas ações** em qualquer estado:

- Basta escolher a ação $ a $ que **maximiza** $ Q(s,a) $.
- Ou seja, a política ótima $ \pi(s) $ é simplesmente:

$$
\pi(s) = \arg\max_{a} Q(s,a)
$$

Agora, a questão é: **como calcular os valores $ Q(s,a) $?**


No **aprendizado por reforço**, existe uma equação fundamental chamada **Equação de Bellman**, que nos ajuda a calcular a função $ Q(s,a) $.

## Definição da Função $ Q(s,a) $

Lembre-se que:

> $ Q(s,a) $ é o retorno esperado se começarmos no estado $ s $, tomarmos a ação $ a $ e depois **seguirmos a melhor política possível**.

Para descrever a **Equação de Bellman**, usaremos a seguinte notação:

### Notação Utilizada

image.png

- **$ S $**: Estado atual.
- **$ R(S) $**: Recompensa associada ao estado $ S $.
  - Exemplo no nosso **MDP**:
    - $ R(1) = 100 $ (Estado 1).
    - $ R(2) = 0 $ (Estado 2).
    - $ R(6) = 40 $ (Estado 6).
- **$ A $**: Ação tomada no estado atual.
- **$ S' $**: Novo estado após tomar a ação $ A $.
  - Exemplo: Se estamos no **estado 4** e tomamos a **ação esquerda**, então chegamos ao **estado 3**.
- **$ A' $**: Ação que pode ser tomada no estado $ S' $.

#### Convenção de Notação:

- $ S, A $ → Estado e ação **atuais**.
- $ S', A' $ → **Próximo** estado e ação.

## A Equação de Bellman

A **Equação de Bellman** é:

$$
Q(s,a) = R(s) + \gamma \max_{a'} Q(s',a')
$$

Onde:

- **$ R(s) $** → Recompensa do estado atual.
- **$ \gamma $** → **Fator de desconto**, que controla a importância de recompensas futuras.
- **$ \max_{a'} Q(s',a') $** → Melhor retorno possível a partir do **próximo estado**.


## Explicação Intuitiva

A equação diz que o **valor de $ Q(s,a) $** é composto por:

1. **A recompensa imediata $ R(s) $**.
2. **O melhor retorno possível do próximo estado $ S' $**, com um fator de desconto $ \gamma $.

Isso significa que a função $ Q(s,a) $ **não é calculada de forma isolada**, mas depende **recursivamente** dos valores futuros de $ Q $.

Essa equação é **fundamental** para os algoritmos de aprendizado por reforço, pois nos permite **calcular iterativamente** os valores de $ Q(s,a) $.

## Aplicando a Equação de Bellman: Exemplos

Vamos aplicar a **Equação de Bellman** a alguns exemplos para calcular os valores de $ Q(s,a) $.

## Exemplo 1: $ Q(2, \text{direita}) $

image.png

Se o **estado atual** for $ S = 2 $ e a **ação escolhida** for **direita**, então o **próximo estado** $ S' $ será o **estado 3**.

A **Equação de Bellman** nos diz que:

$$
Q(2, \text{direita}) = R(2) + \gamma \max_{a'} Q(3, a')
$$

Substituindo os valores conhecidos:

- $ R(2) = 0 $ (a recompensa do estado 2).
- O fator de desconto $ \gamma = 0.5 $.
- Os valores de $ Q(3, a') $ são **25 e 6.25**, então tomamos o **máximo**, que é **25**.

$$
Q(2, \text{direita}) = 0 + 0.5 \times 25
$$

$$
Q(2, \text{direita}) = 12.5
$$

Isso confirma que:

$$
Q(2, \text{direita}) = 12.5
$$

## Exemplo 2: $ Q(4, \text{esquerda}) $

Agora, calculamos $ Q(4, \text{esquerda}) $.

Se o **estado atual** for $ S = 4 $ e a **ação escolhida** for **esquerda**, então o **próximo estado** $ S' $ será **estado 3**.

A **Equação de Bellman** nos dá:

$$
Q(4, \text{esquerda}) = R(4) + \gamma \max_{a'} Q(3, a')
$$

Substituindo os valores conhecidos:

- $ R(4) = 0 $ (a recompensa do estado 4).
- O fator de desconto $ \gamma = 0.5 $.
- Os valores de $ Q(3, a') $ são **25 e 6.25**, então tomamos o **máximo**, que é **25**.

$$
Q(4, \text{esquerda}) = 0 + 0.5 \times 25
$$

$$
Q(4, \text{esquerda}) = 12.5
$$

Isso confirma que:

$$
Q(4, \text{esquerda}) = 12.5
$$

## Casos Especiais: Estados Terminais

Se estamos em um **estado terminal**, a **Equação de Bellman se simplifica**, pois **não há próximo estado $ S' $**.

Nesse caso:

$$
Q(S, A) = R(S)
$$

Ou seja:

- Nos estados terminais, $ Q(s,a) $ **é igual à recompensa do estado**.
- Isso explica por que os valores de $ Q(s,a) $ nos estados terminais são **100 ou 40**.

## Observação sobre Estados Terminais

Se estamos em um **estado terminal**, a **Equação de Bellman** se simplifica para:

$$
Q(s, a) = R(s)
$$

Isso acontece porque **não há um próximo estado $ S' $**, então o **segundo termo desaparece**.

Isso explica por que os valores de $ Q(s, a) $ nos estados terminais são simplesmente **100 ou 40**.


# Revisão: Definição de $ Q(s,a) $

Lembrando que a **função de valor de estado-ação** $ Q(s, a) $ foi definida como:

> $ Q(s,a) $ representa o **retorno esperado** se começarmos no estado $ s $, tomarmos a ação $ a $ e depois **seguirmos a melhor política possível**.

Sabemos que o **melhor retorno possível** de um estado $ s $ é:

$$
\max_{a} Q(s,a)
$$

Podemos renomear $ s $ como $ s' $ e $ a $ como $ a' $, o que nos leva à seguinte forma:

$$
\max_{a'} Q(s', a')
$$

Isso facilita a compreensão da Equação de Bellman.


# Revisando a Equação de Bellman

A **Equação de Bellman** afirma que:

$$
Q(s,a) = R(s) + \gamma \max_{a'} Q(s', a')
$$

### Intuição:

Se começamos no **estado $ s $** e tomamos a **ação $ a $**, então:

1. **Recebemos uma recompensa imediata $ R(s) $**.
2. **Chegamos ao estado $ S' $** e, a partir desse estado, seguimos **a melhor estratégia possível**.
3. **O valor da ação $ a $ no estado $ s $ é a soma da recompensa imediata e do melhor retorno possível a partir de $ S' $, multiplicado pelo fator de desconto $ \gamma $**.


# Explicação Detalhada

O **retorno total** em um problema de **Aprendizado por Reforço** pode ser **decomposto** em duas partes:

1. **A recompensa recebida imediatamente**: $ R(s) $, que também é chamada de **recompensa imediata** na literatura de RL.
2. **O retorno esperado a partir do próximo estado**: Multiplicado pelo fator de desconto $ \gamma $, ou seja:

$$
\gamma \max_{a'} Q(s', a')
$$

Ou seja, se começarmos no estado $ S' $ e agirmos de forma ótima, o **melhor retorno possível** será:

$$
\max_{a'} Q(s', a')
$$

Então, a **Equação de Bellman** nos diz que o **retorno total** pode ser escrito como:

$$
R(s) + \gamma \max_{a'} Q(s', a')
$$


# Outra Maneira de Pensar na Equação de Bellman

Podemos reescrever o **retorno total** como:

$$
R_1 + \gamma R_2 + \gamma^2 R_3 + \gamma^3 R_4 + \dots
$$

Agora, se considerarmos **o que acontece ao partir do estado $ S' $**:

- O **retorno futuro** a partir de $ S' $ será:

$$
R_2 + \gamma R_3 + \gamma^2 R_4 + \dots
$$

Ou seja, **o retorno total a partir de $ S' $ é exatamente o melhor valor de $ Q(s', a') $**.

Se extraímos o fator de desconto $ \gamma $, obtemos:

$$
\gamma \max_{a'} Q(s', a')
$$

Essa é exatamente a **segunda parte da Equação de Bellman**.

## Relacionando com o Exemplo Anterior: $ Q(4, \text{esquerda}) $

Vamos relembrar o cálculo de $ Q(4, \text{esquerda}) $:

- Se começarmos no **estado 4** e tomarmos a ação **esquerda**, então a sequência de recompensas será:
  - **Estado 4** → Recompensa = **0**.
  - **Estado 3** → Recompensa = **0**.
  - **Estado 2** → Recompensa = **0**.
  - **Estado 1 (terminal)** → Recompensa = **100**.

Por isso, o **retorno total** pode ser escrito como:

$$
Q(4, \text{esquerda}) = 0 + 0.5^2 \times 100 + 0.5^3 \times 100
$$

O que resulta em:

$$
Q(4, \text{esquerda}) = 12.5
$$


# Aplicação da Equação de Bellman

A **Equação de Bellman** diz que podemos **decompor** esse cálculo em duas partes:

$$
Q(4, \text{esquerda}) = R(4) + \gamma \max_{a'} Q(3, a')
$$

Substituindo os valores conhecidos:

- $ R(4) = 0 $ (recompensa do estado 4).
- O fator de desconto $ \gamma = 0.5 $.
- O **retorno ótimo** de $ S' = 3 $ é **$ \max_{a'} Q(3, a') = 25 $**.

Então:

$$
Q(4, \text{esquerda}) = 0 + 0.5 \times 25 = 12.5
$$

Ou seja, **a Equação de Bellman nos permite calcular esse valor sem precisar somar toda a sequência de recompensas manualmente**.


# Intuição por Trás da Equação de Bellman

A ideia central da Equação de Bellman é que podemos **quebrar o retorno total** em duas partes:

1. **A recompensa imediata $ R(s) $** (neste caso, **0** no estado 4).
2. **O retorno esperado do próximo estado $ S' $, multiplicado pelo fator de desconto $ \gamma $**.

$$
Q(s, a) = R(s) + \gamma \max_{a'} Q(s', a')
$$

Isso significa que:

- O **retorno total** depende **tanto do que ganhamos agora quanto do que podemos ganhar no futuro**.
- Esse conceito é **fundamental** para o **Aprendizado por Reforço**.

Se essa equação parecer complexa, **não se preocupe**! O mais importante é entender a **intuição geral**:

> **O retorno total pode ser dividido na recompensa imediata mais o retorno futuro esperado com um fator de desconto.**

Mesmo sem compreender todos os detalhes matemáticos, **você pode aplicar a Equação de Bellman corretamente em um algoritmo de aprendizado por reforço**.


# Random (stochastic) environment

Em algumas aplicações, quando tomamos uma **ação**, o **resultado** pode não ser completamente **confiável**.  

Por exemplo, se ordenarmos que um **rover em Marte** vá para a **esquerda**, pode haver:

- Um pequeno deslizamento de pedras.
- Um solo escorregadio, fazendo-o escorregar na direção errada.
- Outros fatores ambientais, como vento forte, que podem desviar sua trajetória.

Na prática, **muitos robôs não conseguem executar exatamente o que foi comandado** devido a fatores imprevisíveis.  

Para lidar com isso, existe uma **generalização do aprendizado por reforço** que modela **ambientes aleatórios ou estocásticos**.

## Exemplo: Rover em um Ambiente Estocástico

Imagine que queremos mover o **rover** para a **esquerda**.

image.png

- **90% das vezes** ($0.9$), ele se move corretamente para a **esquerda**.
- **10% das vezes** ($0.1$), ele **escorrega e vai para a direita**.

Se estivermos no **Estado 4** e comandarmos o rover para **esquerda**:

- **90%** de chance de acabar no **Estado 3**.
- **10%** de chance de acabar no **Estado 5**.

Por outro lado, se ordenarmos que ele vá **para a direita**:

- **90%** de chance de acabar no **Estado 5**.
- **10%** de chance de acabar no **Estado 3**.

Isso define um **ambiente estocástico**, onde as **ações não têm um resultado fixo**, mas sim **distribuições de probabilidade sobre os estados futuros**.

## Impacto no Aprendizado por Reforço

image.png

Suponha que utilizemos a seguinte **política**:

- **Ir para a esquerda** nos estados **2, 3 e 4**.
- **Ir para a direita** no estado **5**.

Agora, vejamos o que pode acontecer ao **seguir essa política** a partir do **Estado 4**.

### **Caso 1: Sucesso total**
- O rover segue **esquerda corretamente**.
- A sequência de estados seria: **4 → 3 → 2 → 1**.
- A sequência de **recompensas**: **0, 0, 0, 100**.

### **Caso 2: Pequeno desvio**
- O rover tenta **ir para a esquerda** no Estado **4**.
- **Mas ele escorrega e volta ao Estado 4**.
- Depois de algumas tentativas, ele consegue seguir **4 → 3 → 2 → 1**.
- A sequência de **recompensas** pode ser: **0, 0, 0, 0, 100**.

### **Caso 3: Desvio completo**
- O rover tenta **ir para a esquerda**, mas **escorrega e vai para a direita (Estado 5)**.
- Em **Estado 5**, ele tenta **ir para a direita** e chega ao **Estado 6**.
- A sequência de **recompensas**: **0, 0, 40**.

Em **ambientes estocásticos**, **o mesmo conjunto de ações pode levar a diferentes sequências de recompensas**.


## **Maximizando o Retorno Esperado**

Em problemas determinísticos, podíamos calcular o **retorno** como:

$$
R = R_1 + \gamma R_2 + \gamma^2 R_3 + \dots
$$

No entanto, **em um ambiente estocástico**, **não há apenas uma única sequência de recompensas**.

- Cada **execução** da política pode gerar **uma sequência diferente** de recompensas.
- O **retorno total** agora é **aleatório**.

Portanto, **não queremos maximizar apenas um único retorno**, mas sim **a média dos retornos ao longo de várias execuções**.

Isso nos leva ao conceito de **retorno esperado**.


## **O Retorno Esperado**

Para lidar com essa aleatoriedade, em Aprendizado por Reforço buscamos **maximizar o valor esperado** do retorno:

$$
E[R_1 + \gamma R_2 + \gamma^2 R_3 + \dots]
$$

- **$ E $** representa o **valor esperado** (ou média estatística).
- **Queremos encontrar uma política $ \pi $ que maximize o valor médio do retorno**.

Na estatística, o termo **"esperado"** significa simplesmente **média**.  

Isso quer dizer que, se rodássemos a política $ \pi $ **milhares de vezes**, pegaríamos a **média dos retornos obtidos**.

Quando temos um **problema de Aprendizado por Reforço Estocástico** ou um **Processo de Decisão de Markov Estocástico (MDP Estocástico)**, nosso objetivo é:

> **Escolher uma política $ \pi(s) $ que maximize o retorno esperado**.

Isso significa que, ao invés de **maximizar um único retorno**, buscamos **maximizar a média dos retornos possíveis** devido à aleatoriedade no ambiente.

## **Modificação da Equação de Bellman**

A **Equação de Bellman** continua sendo fundamental, mas agora precisa ser **ajustada** para lidar com a **aleatoriedade** dos estados futuros.

Aqui está a Equação de Bellman tradicional:

$$
Q(s,a) = R(s) + \gamma \max_{a'} Q(s', a')
$$

A **diferença principal** em ambientes estocásticos é que, ao tomarmos a ação $ a $ no estado $ s $, o **próximo estado $ s' $ é aleatório**.

Por exemplo:

- Se o rover está no **Estado 3** e tenta ir para a **esquerda**:
  - Ele pode acabar no **Estado 2**.
  - Ou pode escorregar e acabar no **Estado 4**.

Isso significa que **$ s' $ não é fixo**, mas segue uma **distribuição de probabilidade**.

### **Bellman Equation para MDPs Estocásticos**
Agora, a equação de Bellman se torna:

$$
Q(s,a) = R(s) + \gamma \mathbb{E}_{s'} \left[ \max_{a'} Q(s', a') \right]
$$

Onde:

- **$ \mathbb{E}_{s'} $** indica o **valor esperado** considerando **todas as possibilidades** para $ s' $.
- **Ao invés de calcular um único valor de $ Q(s', a') $**, agora fazemos uma **média ponderada** dos valores possíveis, considerando as probabilidades de cada $ s' $.

# Example of continuous state space applications

Muitas aplicações de controle robótico possuem espaços de estado contínuos. Vamos analisar o que isso significa e como generalizar o conceito que discutimos para esses espaços de estado contínuos.

image.png

No exemplo simplificado de um rover em Marte que usamos, consideramos um conjunto discreto de estados. Isso significa que o rover simplificado de Marte poderia estar em apenas uma de seis posições possíveis. No entanto, a maioria dos robôs pode ocupar mais do que seis posições ou qualquer número discreto de posições. Em vez disso, eles podem estar em qualquer uma de um número muito grande de posições com valores contínuos.

Por exemplo, se o rover de Marte pudesse estar em qualquer ponto ao longo de uma linha, sua posição poderia ser representada por um número variando de 0 a 6 quilômetros, onde qualquer número intermediário seria válido. Isso seria um exemplo de um espaço de estado contínuo, porque a posição seria representada por um número como 2,7 quilômetros, 4,8 quilômetros ou qualquer outro valor entre zero e seis.

### Exemplo de Controle de Carro ou Caminhão

Vamos analisar outro exemplo: o controle de um carro ou caminhão. Aqui está um caminhão de brinquedo. Se você estiver desenvolvendo um carro ou caminhão autônomo e quiser controlá-lo para dirigir suavemente, o estado desse caminhão pode incluir alguns números, tais como:

- Sua posição em **x** e **y**;
- Sua orientação (para qual direção está virado).

Supondo que o caminhão permaneça no solo, provavelmente não precisamos nos preocupar com sua altura. Assim, o estado incluirá **x**, **y** e um ângulo **θ** (Theta) representando sua orientação. Além disso, podemos incluir:

- Sua velocidade na direção **x**, que denotaremos como **ẋ** (x ponto), indicando a taxa de variação da coordenada **x**;
- Sua velocidade na direção **y**, que denotaremos como **ẏ** (y ponto);
- A taxa de variação do ângulo **θ**, que chamaremos de **θ̇** (Theta ponto), ou seja, a velocidade de rotação do carro.

image.png

Enquanto no exemplo do rover de Marte o estado era representado por um único número entre seis possíveis (1, 2, 3, 4, 5 ou 6), no caso do carro, o estado seria composto por um vetor de seis números. Cada um desses números pode assumir qualquer valor dentro de um intervalo válido. Por exemplo, **θ** deve variar entre 0 e 360 graus.

### Exemplo de Controle de um Helicóptero Autônomo

Agora, vamos considerar um helicóptero autônomo. Como caracterizaríamos sua posição? Para ilustrar, imagine um pequeno helicóptero de brinquedo. A posição do helicóptero incluiria:

- **x**: Sua posição no eixo norte-sul;
- **y**: Sua posição no eixo leste-oeste;
- **z**: Sua altura em relação ao solo.

Além da posição, o helicóptero também possui uma orientação. Convencionalmente, podemos capturar essa orientação com três números adicionais:

1. **Rolagem (roll)**: Indica se o helicóptero está inclinando-se para a esquerda ou para a direita;
2. **Inclinação (pitch)**: Indica se o helicóptero está inclinando-se para frente ou para trás;
3. **Guinada (yaw)**: Indica sua orientação em relação à bússola (se está apontando para o norte, leste, sul ou oeste).

image.png

Para resumir, o estado do helicóptero inclui:

- **Posição:** (**x**, **y**, **z**)  
- **Orientação:** (**Φ**, **θ**, **ψ**)  
  - Onde **Φ** (Phi) representa a rolagem,
  - **θ** (Theta) representa a inclinação,
  - **ψ** (Psi) representa a guinada.

Mas para controlar o helicóptero, também precisamos saber sua velocidade em cada uma dessas direções:

- **Velocidade em x** (denotada por **ẋ**)  
- **Velocidade em y** (denotada por **ẏ**)  
- **Velocidade em z** (denotada por **ż**)  

Além disso, precisamos considerar a velocidade angular, ou seja, quão rapidamente a orientação do helicóptero está mudando:

- **Taxa de variação da rolagem (roll̇)**
- **Taxa de variação da inclinação (pitcḣ)**
- **Taxa de variação da guinada (yaẇ)**

Esse conjunto de 12 números representa o estado usado para controlar helicópteros autônomos. Essa lista de valores é utilizada como entrada para uma **política de controle**. O trabalho dessa política é analisar esses 12 números e decidir qual ação tomar para controlar o helicóptero adequadamente.

### Estados Contínuos em Aprendizado por Reforço

Nos problemas de aprendizado por reforço com estados contínuos, ou em um **processo de decisão de Markov com estados contínuos (continuous-state MDP)**, o estado do problema não é apenas um valor discreto entre um pequeno conjunto de opções, como um número de 1 a 6. Em vez disso, é um vetor de números, cada um dos quais pode assumir um grande conjunto de valores possíveis.

# Lunar lander

O módulo lunar permite que você controle um veículo simulado para pousar na Lua. É como um pequeno e divertido jogo que tem sido amplamente utilizado por pesquisadores de aprendizado por reforço. Vamos dar uma olhada no que ele é.

### O Objetivo do Jogo

image.png

Nesta aplicação, você está no comando de um módulo lunar que está se aproximando rapidamente da superfície da Lua. Seu trabalho é acionar os propulsores nos momentos adequados para pousá-lo com segurança na plataforma de pouso.  

Para dar uma ideia de como isso funciona, aqui está um módulo lunar aterrissando com sucesso. Ele dispara propulsores para baixo e para os lados (esquerda e direita) para se posicionar corretamente e pousar entre as duas bandeiras amarelas.  

image.png

Por outro lado, se a política de aprendizado por reforço não for eficaz, isso pode resultar em um pouso mal-sucedido, onde o módulo acaba se espatifando na superfície lunar.

### Ações Possíveis

image.png

Nesta aplicação, você tem quatro possíveis ações a cada intervalo de tempo:

1. **Não fazer nada**: Se nenhuma ação for tomada, a força da inércia e da gravidade puxará o módulo em direção à superfície lunar.
2. **Acionar o propulsor esquerdo**: Isso é indicado por um pequeno ponto vermelho à esquerda. Acionar este propulsor empurrará o módulo para a direita.
3. **Acionar o propulsor principal**: Ele dispara para baixo e gera um empuxo para reduzir a velocidade de queda.
4. **Acionar o propulsor direito**: Isso empurra o módulo para a esquerda.

Seu trabalho é escolher continuamente essas ações para garantir que o módulo lunar pouse com segurança entre as bandeiras na plataforma de pouso.

Para simplificar os nomes das ações, usaremos as seguintes abreviações:
- **Nada**: Significa não fazer nada.
- **Esquerda**: Acionar o propulsor esquerdo.
- **Principal**: Acionar o propulsor principal.
- **Direita**: Acionar o propulsor direito.

Esses nomes curtos serão usados mais adiante.

### O Espaço de Estados

O espaço de estados do módulo lunar é representado por várias variáveis:

- **Posição** `(x, y)`: Indica a posição do módulo no eixo horizontal (`x`) e no eixo vertical (`y`).
- **Velocidade** `(vx, vy)`: Representa a velocidade do módulo nos eixos horizontal e vertical.
- **Ângulo** `(θ)`: Indica o quanto o módulo está inclinado para a esquerda ou para a direita.
- **Velocidade angular** `(θ̇)`: Mede a taxa de variação da inclinação do módulo.
- **Contato com o solo** `(l, r)`: São duas variáveis binárias que indicam se as pernas esquerda (`l`) e direita (`r`) do módulo estão tocando o solo (valores `0` ou `1`).

Como pequenas diferenças na posição podem impactar significativamente se o módulo pousou ou não, essas variáveis são essenciais para a simulação.

### A Função de Recompensa

A função de recompensa do módulo lunar determina os pontos atribuídos com base no desempenho do pouso. Os critérios são:

1. **Pouso bem-sucedido**: Se o módulo alcançar a plataforma de pouso, ele recebe uma recompensa entre **100 e 140**, dependendo da precisão do pouso no centro da plataforma.
2. **Movimento em direção à plataforma**: Se o módulo se move para mais perto da plataforma, recebe uma recompensa positiva; se se afasta, recebe uma recompensa negativa.
3. **Colisão**: Se o módulo colidir com a superfície lunar, recebe uma grande penalização de **-100**.
4. **Pouso suave**: Se o pouso for bem-sucedido sem ser uma colisão forte, o módulo recebe um bônus adicional de **+100**.
5. **Contato das pernas com o solo**: Para cada perna do módulo que toca o solo de forma estável, ele recebe **+10** pontos.
6. **Uso de combustível**: Para evitar desperdício de combustível, há penalizações:
   - Acionar o propulsor principal gera uma penalidade de **-0.3** por cada acionamento.
   - Acionar os propulsores laterais gera uma penalidade de **-0.03** por cada acionamento.

Essa função de recompensa foi cuidadosamente projetada para incentivar comportamentos desejáveis (pouso controlado e eficiente) e desencorajar comportamentos indesejáveis (colisões e desperdício de combustível).

Ao projetar uma aplicação de aprendizado por reforço, normalmente leva algum tempo para definir exatamente quais comportamentos queremos incentivar ou desencorajar e traduzi-los em uma função de recompensa adequada. No entanto, mesmo que essa definição exija esforço, ela ainda é muito mais fácil do que especificar manualmente a ação correta para cada estado possível do módulo lunar, algo extremamente complexo.

### O Problema do Módulo Lunar no Aprendizado por Reforço

O objetivo do problema do módulo lunar no aprendizado por reforço é aprender uma **política** (π) que, dado um estado **S**, escolha uma ação **A = π(S)** que maximize o **retorno**, ou seja, a soma dos valores de recompensa ao longo do tempo, considerando um fator de desconto.

Para este problema, costuma-se utilizar um fator de desconto **γ = 0.985**, que é bastante alto, o que significa que recompensas futuras ainda são valorizadas significativamente.

Se conseguirmos aprender uma política **π** que maximize o retorno esperado, então teremos um agente capaz de pousar o módulo lunar com sucesso.

# Learning the state-value function

Vamos ver como podemos usar o aprendizado por reforço para controlar o **Lunar Lander** ou resolver outros problemas de aprendizado por reforço. A ideia principal é treinar uma **rede neural** para calcular ou aproximar a **função de valor de estado-ação** $ Q(s, a) $, que, por sua vez, nos permitirá escolher boas ações. Vamos entender como isso funciona.

## Estrutura do Algoritmo de Aprendizado

image.png

O coração do algoritmo de aprendizado está no treinamento de uma **rede neural** que recebe como entrada o estado atual e a ação atual e calcula ou aproxima $ Q(s, a) $.

No caso do **Lunar Lander**, tomamos o **estado** $ s $ e qualquer **ação** $ a $ e os combinamos. Concretamente, o **estado** é representado por uma lista de **oito números**, que são:

- **$ x, y $** → Coordenadas da posição da nave.
- **$ \dot{x}, \dot{y} $** → Velocidades nos eixos $ x $ e $ y $.
- **$ \theta, \dot{\theta} $** → Ângulo e taxa de variação do ângulo.
- **$ L, R $** → Indicadores de contato das pernas com o solo (perna esquerda e direita).

Além disso, há **quatro possíveis ações**:

1. **Nada** (sem propulsão)
2. **Propulsor esquerdo**
3. **Propulsor principal**
4. **Propulsor direito**

Para representar essas ações, utilizamos um **vetor de características one-hot**. Por exemplo:

- Se a ação for **nada**, usamos $ [1, 0, 0, 0] $.
- Se a ação for **propulsor esquerdo**, usamos $ [0, 1, 0, 0] $.

Assim, a entrada para a **rede neural** será um vetor de **12 números**: os **8 números do estado** e os **4 números da codificação one-hot da ação**. Chamaremos essa entrada de **$ X $**.

A rede neural terá a seguinte estrutura:

- **Camada de entrada:** 12 neurônios (representando $ X $).
- **Primeira camada oculta:** 64 neurônios.
- **Segunda camada oculta:** 64 neurônios.
- **Camada de saída:** 1 neurônio (calculando $ Q(s, a) $).

O objetivo da rede neural é estimar a **função de valor de estado-ação** para o Lunar Lander, dado um **estado** $ s $ e uma **ação** $ a $. Essa estimativa será chamada de **valor alvo $ Y $**, o qual será usado para treinar a rede neural.

## Diferença entre Aprendizado por Reforço e Aprendizado Supervisionado

Embora o **aprendizado por reforço** seja diferente do **aprendizado supervisionado**, o que faremos será parecido com um treinamento supervisionado:

- Não vamos treinar a rede para **mapear estados diretamente para ações**.
- Em vez disso, treinamos a rede para **mapear pares estado-ação para $ Q(s, a) $**.

Esse método se baseia em redes neurais dentro do **algoritmo de aprendizado por reforço**, o que geralmente funciona bem. Se conseguirmos **treinar a rede corretamente**, então, sempre que o Lunar Lander estiver em um **estado $ s $**, podemos calcular os valores $ Q(s, a) $ para as **quatro ações possíveis**:

- $ Q(s, \text{nada}) $
- $ Q(s, \text{esquerda}) $
- $ Q(s, \text{principal}) $
- $ Q(s, \text{direita}) $

A **ação escolhida** será aquela que tiver o maior valor $ Q(s, a) $. Por exemplo, se $ Q(s, \text{principal}) $ for o maior, a nave acionará o **propulsor principal**.

## Como Treinar a Rede Neural para Calcular $ Q(s, a) $?

A abordagem utilizada será baseada na **Equação de Bellman**:

$$
Q(s, a) = R(s) + \gamma \max_{a'} Q(s', a')
$$

Onde:

- $ R(s) $ é a **recompensa** por estar no estado $ s $.
- $ \gamma $ é o **fator de desconto**.
- $ \max_{a'} Q(s', a') $ representa a **melhor estimativa futura** da função $ Q $.

O lado direito da equação nos dá o valor desejado para $ Q(s, a) $, que chamamos de **$ Y $**, enquanto a entrada para a rede é o par $ (s, a) $, que chamamos de **$ X $**.

Agora surge a pergunta: **como obtemos um conjunto de treinamento de pares $ (X, Y) $?**

A resposta é **coletar experiências de interação com o ambiente**.

## Coletando Dados para Treinamento

Para treinar a rede, seguimos estes passos:

1. **Executamos o Lunar Lander com ações aleatórias**.
   - Ligamos os propulsores aleatoriamente.
   - Coletamos experiências boas e ruins.

2. **Salvamos as experiências como tuplas $ (s, a, R, s') $**.
   - $ s $ → Estado atual.
   - $ a $ → Ação tomada.
   - $ R $ → Recompensa recebida.
   - $ s' $ → Novo estado alcançado.

3. **Geramos pares de treinamento $ (X, Y) $**:
   - $ X = (s, a) $.
   - $ Y = R(s) + \gamma \max_{a'} Q(s', a') $.

Isso nos dá **10.000 exemplos** de treinamento.

Inicialmente, a função $ Q $ será um **palpite aleatório**, mas melhorará com o tempo.

Aqui está o texto traduzido, completo e estruturado:

## Como Obtemos $ Q(s', a') $?

Agora, você pode estar se perguntando: **de onde vem $ Q(s', a') $ ou $ Q(s'^1, a') $?**

Inicialmente, **não sabemos qual é a função $ Q $**. Mas, quando não sabemos qual é a função $ Q $, podemos **começar com um palpite totalmente aleatório**.

No início do treinamento, simplesmente assumimos **valores aleatórios** para $ Q(s, a) $. Com o tempo, esses valores vão sendo refinados conforme o algoritmo aprende.

### Exemplo de Geração de Dados de Treinamento

Vamos analisar um segundo exemplo:

- Suponha que você esteja em um **estado $ S^2 $**.
- Você tomou uma **ação $ a^2 $**.
- Você recebeu uma **recompensa $ R(S^2) $**.
- Como resultado, você chegou a um **novo estado $ S'^2 $**.

A partir dessa experiência, podemos criar um **segundo exemplo de treinamento** no nosso conjunto de dados:

- A entrada será $ X^2 = (S^2, a^2) $, ou seja, **o estado e a ação**.
- O valor alvo $ Y^2 $ será calculado pela **equação de Bellman**:

$$
Y^2 = R(S^2) + \gamma \max_{a'} Q(S'^2, a')
$$

O valor $ Q(S'^2, a') $ pode ser **inicialmente uma estimativa aleatória**, mas será refinado ao longo do treinamento.

Este processo continua, gerando novos exemplos até termos **10.000 pares de treinamento $ (X, Y) $**.


## Criando o Conjunto de Treinamento para a Rede Neural

Agora que coletamos **10.000 exemplos de treinamento**, podemos treinar uma rede neural.

- As **entradas** ($ X $) serão **vetores de 12 números**, compostos pelos **8 números do estado** e pelos **4 números da ação (one-hot encoding)**.
- Os **valores-alvo** ($ Y $) serão **valores escalares** obtidos pela equação de Bellman.

A rede neural será treinada para **minimizar o erro quadrático médio (MSE)**, ou seja, para prever corretamente $ Y $ a partir de $ X $.

Isso é **apenas uma parte do algoritmo de aprendizado**. Agora, vamos ver como ele funciona como um todo.


## O Algoritmo Completo para Aprendizado da Função $ Q $

Agora, vamos montar o **algoritmo completo para aprender a função $ Q $**.

- **Inicialização da Rede Neural**  
   - Definimos uma **rede neural** com parâmetros **inicialmente aleatórios**.
   - No começo, **não temos nenhuma estimativa válida da função $ Q $**.
   - Isso é semelhante ao treinamento de uma **regressão linear**, onde os parâmetros são inicializados aleatoriamente antes de serem ajustados pelo **gradiente descendente**.

- **Execução de Ações no Lunar Lander**  
   - O agente começa a **executar ações aleatórias** no ambiente.
   - Ele **coleta experiências** registrando tuplas $ (s, a, R, s') $.
   - Mantemos na memória **apenas as 10.000 experiências mais recentes** (**Replay Buffer**).

- **Treinamento Periódico da Rede Neural**  
   - Periodicamente, usamos os **10.000 exemplos armazenados** para treinar a rede.
   - O conjunto de treinamento será formado por:
     - **Entradas $ X $ → (estado, ação)**
     - **Saídas $ Y $ → $ R(s) + \gamma \max_{a'} Q(s', a') $**
   - A rede é treinada para **minimizar a diferença entre a previsão $ Q(s, a) $ e $ Y $**.

- **Melhoria Contínua da Estimativa de $ Q $**  
   - Após o treinamento, **substituímos a função $ Q $ antiga pela nova rede neural treinada**.
   - Esse processo é repetido indefinidamente.
   - Com o tempo, a estimativa de $ Q(s, a) $ melhora.


## O Algoritmo **DQN** (Deep Q-Network)

O algoritmo que acabamos de descrever é chamado **DQN** (**Deep Q-Network**), pois utiliza **redes neurais profundas** para aprender a função $ Q(s, a) $.

Esse método foi desenvolvido por **Mnih et al.** e se mostrou eficaz para resolver muitos problemas de aprendizado por reforço.

Se utilizarmos esse algoritmo exatamente como descrito, ele **funcionará razoavelmente bem no Lunar Lander**. No entanto:

- Pode **demorar muito tempo para convergir**.
- Talvez **não atinja uma aterrissagem perfeita**.

### Melhorias no Algoritmo

Com algumas **otimizações**, o DQN pode se tornar **muito mais eficiente**, como:

- **Exploração e Exploração Balanceadas ($\epsilon$-Greedy)**  
   - Permite ao agente **explorar novas ações** enquanto **mantém escolhas eficazes**.

- **Redes-Alvo para Estabilização**  
   - Em vez de atualizar $ Q(s, a) $ imediatamente, utilizamos **uma cópia estável da rede** para calcular os valores futuros.

- **Replay Prioritário**  
   - Algumas experiências são **mais importantes** para o aprendizado e devem ser priorizadas no treinamento.

Essas técnicas fazem com que o DQN aprenda **mais rápido e com mais estabilidade**.


# Algorithm refinement: Improved neural network architecture

Vimos uma **arquitetura de rede neural** que recebe como entrada o **estado e a ação** e tenta **calcular a função $ Q(s, a) $**. No entanto, existe uma **modificação** nessa arquitetura que torna o **algoritmo muito mais eficiente**.

Na maioria das implementações do **Deep Q-Network (DQN)**, essa versão **mais otimizada** da arquitetura é utilizada. Vamos entender como ela funciona.

### **Arquitetura Original da Rede Neural**
Anteriormente, a rede neural funcionava da seguinte maneira:

image.png

**Entrada:**  
   - Um vetor de **12 números**:  
     - **8 números** representando o estado $ s $ do Lunar Lander.  
     - **4 números** (one-hot encoding) representando a ação $ a $.  

**Saída:**  
   - Um **único valor $ Q(s, a) $** correspondente ao par **(estado, ação)** inserido.

**Problema:**  
   - Para escolher a **melhor ação**, precisaríamos **rodar a inferência da rede 4 vezes** (uma vez para cada possível ação: **nada, esquerda, principal e direita**).
   - Isso torna o processo **ineficiente**, pois cada estado exige **4 avaliações separadas** da rede.


### **Nova Arquitetura Mais Eficiente**
A solução para melhorar a eficiência é **modificar a rede para calcular os quatro valores $ Q(s, a) $ ao mesmo tempo**.

image.png

**Entrada:**  
   - Apenas os **8 números do estado $ s $** (sem precisar da codificação one-hot da ação).

**Arquitetura Interna:**  
   - Primeira camada oculta com **64 neurônios**.
   - Segunda camada oculta com **64 neurônios**.

**Saída:**  
   - **4 neurônios na camada de saída**, onde cada neurônio corresponde a um valor $ Q(s, a) $ para cada uma das **4 ações possíveis**:
     - $ Q(s, \text{nada}) $
     - $ Q(s, \text{esquerda}) $
     - $ Q(s, \text{principal}) $
     - $ Q(s, \text{direita}) $


### **Vantagens da Nova Arquitetura**
**Inferência Mais Rápida**
   - Em vez de rodar a rede **4 vezes** (uma para cada ação), agora **basta uma única inferência** para obter os valores $ Q(s, a) $ de todas as ações.
   - Isso permite que o agente **escolha rapidamente** a melhor ação $ a $ apenas pegando o valor máximo da saída da rede.

**Otimização na Equação de Bellman**
   - A Equação de Bellman exige calcular:

     $$
     \max_{a'} Q(s', a')
     $$

   - Com a nova arquitetura, **todos os $ Q(s', a') $ já estão disponíveis** ao mesmo tempo.
   - Assim, **o valor máximo pode ser extraído instantaneamente**, tornando o cálculo **mais eficiente**.

# Algorithm refinement: ϵ-greedy policy

O algoritmo de aprendizado que desenvolvemos, mesmo enquanto você ainda está aprendendo como aproximar Q(s,a), requer que você tome algumas ações no Lunar Lander. Como escolher essas ações enquanto ainda está aprendendo? A maneira mais comum é utilizar algo chamado política epsilon-greedy. Vejamos como isso funciona.

image.png

Aqui está o algoritmo que você viu anteriormente. Uma das etapas do algoritmo é tomar ações no Lunar Lander. Quando o algoritmo ainda está em execução, não sabemos realmente qual é a melhor ação a ser tomada em cada estado. Se soubéssemos, já teríamos terminado o aprendizado. Mas mesmo enquanto ainda estamos aprendendo e não temos uma boa estimativa de Q(s,a), como escolhemos ações nessa etapa do algoritmo de aprendizado? Vamos considerar algumas opções.

image.png

Quando você está em algum estado s, talvez não queira tomar ações completamente ao acaso, pois frequentemente isso resultará em uma ação ruim. Uma opção natural seria, sempre que estiver no estado s, escolher a ação a que maximiza Q(s,a). Poderíamos dizer que, mesmo que Q(s,a) ainda não seja uma ótima estimativa da função Q, vamos fazer o nosso melhor e usar nosso palpite atual para Q(s,a), escolhendo a ação que o maximiza. Acontece que isso pode funcionar razoavelmente, mas não é a melhor opção.

Em vez disso, o que é comumente feito é a seguinte opção (opção 2): na maior parte do tempo, digamos com uma probabilidade de 0,95, escolha a ação que maximiza Q(s,a). Ou seja, na maioria das vezes tentamos escolher uma boa ação usando nossa estimativa atual de Q(s,a). Porém, em uma pequena fração do tempo, digamos cinco por cento das vezes, escolhemos uma ação aleatoriamente.

Por que queremos ocasionalmente escolher uma ação aleatória? Aqui está o motivo. Suponha que, por alguma razão estranha, Q(s,a) foi inicializado aleatoriamente, de forma que o algoritmo de aprendizado acredita que acionar o propulsor principal nunca seja uma boa ideia. Talvez os parâmetros da rede neural tenham sido inicializados de forma que Q(s, principal) seja sempre muito baixo. Nesse caso, como a rede neural está tentando escolher a ação a que maximiza Q(s,a), ela nunca tentará acionar o propulsor principal. E como nunca tentará acionar o propulsor principal, ela nunca aprenderá que acionar o propulsor principal às vezes é realmente uma boa ideia.

Devido à inicialização aleatória, se a rede neural inicialmente, por acaso, decidir que algumas ações são ruins, então, na opção 1, nunca experimentará essas ações para descobrir que, na verdade, podem ser boas ações, como acionar o propulsor principal. Já na opção 2, a cada passo temos uma pequena probabilidade de tentar ações diferentes, permitindo que a rede neural aprenda a superar possíveis preconceitos iniciais sobre ações que podem parecer ruins, mas que na realidade não o são.

Essa ideia de escolher ações aleatoriamente é às vezes chamada de etapa de exploração, pois experimentaremos algo que talvez não seja a melhor ideia, mas vamos explorar e aprender mais sobre uma ação em circunstâncias nas quais talvez não tenhamos muita experiência.

Escolher uma ação que maximiza Q(s,a) é às vezes chamado de ação "greedy" (gananciosa), pois estamos tentando maximizar nosso retorno ao escolher essa ação. Na literatura de aprendizado por reforço, isso também é referido como etapa de exploração ou exploitation. Sei que "abuso" (exploitation) não é algo bom, ninguém deveria explorar outras pessoas, mas historicamente esse termo foi adotado no aprendizado por reforço para indicar que devemos usar tudo o que aprendemos para obter o melhor desempenho possível.

Na literatura de aprendizado por reforço, às vezes as pessoas falam sobre o trade-off entre exploração e abuso (exploration vs. exploitation), que se refere à frequência com que você escolhe ações aleatoriamente ou ações que talvez não sejam as melhores para aprender mais, versus tentar maximizar seu retorno escolhendo a ação que maximiza Q(s,a).

Essa abordagem, ou seja, a opção 2, tem um nome: é chamada política epsilon-greedy, onde epsilon (por exemplo, 0,05) é a probabilidade de escolher uma ação aleatoriamente. Esta é a maneira mais comum de fazer seu algoritmo de aprendizado por reforço explorar um pouco, mesmo enquanto a maior parte do tempo ou ocasionalmente tomando ações greedy.

Aliás, muitas pessoas comentaram que o nome "política epsilon-greedy" é confuso, já que você na verdade está sendo greedy 95% do tempo, não 5%. Talvez um nome mais preciso fosse "política (1 menos epsilon)-greedy", pois é greedy 95% do tempo e explora apenas 5%. Contudo, por razões históricas, o termo "política epsilon-greedy" prevaleceu e é utilizado para indicar a política que explora com uma fração epsilon do tempo.

Por último, um truque frequentemente usado em aprendizado por reforço é começar com epsilon alto, inicialmente realizando ações aleatórias com bastante frequência, e depois gradualmente diminuir o epsilon com o tempo, tornando cada vez menos provável tomar ações aleatórias e mais provável utilizar estimativas cada vez melhores da função Q para escolher boas ações. Por exemplo, no exercício do Lunar Lander, você pode começar com epsilon muito alto, talvez até epsilon igual a 1, escolhendo ações completamente aleatórias no início e gradualmente diminuindo para 0,01, para que eventualmente você tome ações greedy 99% do tempo e ações aleatórias em apenas 1% das vezes.

Se isso parecer complicado, não se preocupe. Forneceremos o código no Jupyter Lab que mostrará como fazer isso. Caso você implemente o algoritmo conforme descrito, com uma arquitetura de rede neural mais eficiente e uma política epsilon-greedy de exploração, perceberá que funciona bastante bem no Lunar Lander.

Uma observação importante é que os algoritmos de aprendizado por reforço são mais sensíveis à escolha de hiperparâmetros, especialmente em comparação com aprendizado supervisionado. Por exemplo, no aprendizado supervisionado, se você ajustar a taxa de aprendizado para um valor um pouco menor, o algoritmo pode demorar um pouco mais para aprender, talvez três vezes mais, o que pode ser chato, mas tolerável. Já no aprendizado por reforço, se você escolher um valor incorreto para epsilon ou outros parâmetros, o tempo necessário para aprendizado pode aumentar em 10 ou até 100 vezes. Algoritmos de aprendizado por reforço, talvez por serem menos maduros, são muito mais sensíveis a pequenas escolhas de parâmetros, sendo frequentemente mais frustrantes de ajustar do que algoritmos de aprendizado supervisionado.

````python
"""
Utilities module containing helper functions for the Deep Q-Learning - Lunar Lander
Jupyter notebook (C3_W3_A1_Assignment) from DeepLearning.AI's "Unsupervised Learning,
Recommenders, Reinforcement Learning" course on Coursera.
"""

import base64
import random
from itertools import zip_longest

import imageio
import IPython
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
import pandas as pd
import tensorflow as tf

SEED = 0  # Seed for the pseudo-random number generator.
MINIBATCH_SIZE = 64  # Mini-batch size.
TAU = 1e-3  # Soft update parameter.
E_DECAY = 0.995  # ε-decay rate for the ε-greedy policy.
E_MIN = 0.01  # Minimum ε value for the ε-greedy policy.


random.seed(SEED)


def get_experiences(memory_buffer):
    """
    Returns a random sample of experience tuples drawn from the memory buffer.

    Retrieves a random sample of experience tuples from the given memory_buffer and
    returns them as TensorFlow Tensors. The size of the random sample is determined by
    the mini-batch size (MINIBATCH_SIZE).

    Args:
        memory_buffer (deque):
            A deque containing experiences. The experiences are stored in the memory
            buffer as namedtuples: namedtuple("Experience", field_names=["state",
            "action", "reward", "next_state", "done"]).

    Returns:
        A tuple (states, actions, rewards, next_states, done_vals) where:

            - states are the starting states of the agent.
            - actions are the actions taken by the agent from the starting states.
            - rewards are the rewards received by the agent after taking the actions.
            - next_states are the new states of the agent after taking the actions.
            - done_vals are the boolean values indicating if the episode ended.

        All tuple elements are TensorFlow Tensors whose shape is determined by the
        mini-batch size and the given Gym environment. For the Lunar Lander environment
        the states and next_states will have a shape of [MINIBATCH_SIZE, 8] while the
        actions, rewards, and done_vals will have a shape of [MINIBATCH_SIZE]. All
        TensorFlow Tensors have elements with dtype=tf.float32.
    """

    experiences = random.sample(memory_buffer, k=MINIBATCH_SIZE)
    states = tf.convert_to_tensor(
        np.array([e.state for e in experiences if e is not None]), dtype=tf.float32
    )
    actions = tf.convert_to_tensor(
        np.array([e.action for e in experiences if e is not None]), dtype=tf.float32
    )
    rewards = tf.convert_to_tensor(
        np.array([e.reward for e in experiences if e is not None]), dtype=tf.float32
    )
    next_states = tf.convert_to_tensor(
        np.array([e.next_state for e in experiences if e is not None]), dtype=tf.float32
    )
    done_vals = tf.convert_to_tensor(
        np.array([e.done for e in experiences if e is not None]).astype(np.uint8),
        dtype=tf.float32,
    )
    return (states, actions, rewards, next_states, done_vals)


def check_update_conditions(t, num_steps_upd, memory_buffer):
    """
    Determines if the conditions are met to perform a learning update.

    Checks if the current time step t is a multiple of num_steps_upd and if the
    memory_buffer has enough experience tuples to fill a mini-batch (for example, if the
    mini-batch size is 64, then the memory buffer should have more than 64 experience
    tuples in order to perform a learning update).

    Args:
        t (int):
            The current time step.
        num_steps_upd (int):
            The number of time steps used to determine how often to perform a learning
            update. A learning update is only performed every num_steps_upd time steps.
        memory_buffer (deque):
            A deque containing experiences. The experiences are stored in the memory
            buffer as namedtuples: namedtuple("Experience", field_names=["state",
            "action", "reward", "next_state", "done"]).

    Returns:
       A boolean that will be True if conditions are met and False otherwise.
    """

    if (t + 1) % num_steps_upd == 0 and len(memory_buffer) > MINIBATCH_SIZE:
        return True
    else:
        return False


def get_new_eps(epsilon):
    """
    Updates the epsilon value for the ε-greedy policy.

    Gradually decreases the value of epsilon towards a minimum value (E_MIN) using the
    given ε-decay rate (E_DECAY).

    Args:
        epsilon (float):
            The current value of epsilon.

    Returns:
       A float with the updated value of epsilon.
    """

    return max(E_MIN, E_DECAY * epsilon)


def get_action(q_values, epsilon=0.0):
    """
    Returns an action using an ε-greedy policy.

    This function will return an action according to the following rules:
        - With probability epsilon, it will return an action chosen at random.
        - With probability (1 - epsilon), it will return the action that yields the
        maximum Q value in q_values.

    Args:
        q_values (tf.Tensor):
            The Q values returned by the Q-Network. For the Lunar Lander environment
            this TensorFlow Tensor should have a shape of [1, 4] and its elements should
            have dtype=tf.float32.
        epsilon (float):
            The current value of epsilon.

    Returns:
       An action (numpy.int64). For the Lunar Lander environment, actions are
       represented by integers in the closed interval [0,3].
    """

    if random.random() > epsilon:
        return np.argmax(q_values.numpy()[0])
    else:
        return random.choice(np.arange(4))


def update_target_network(q_network, target_q_network):
    """
    Updates the weights of the target Q-Network using a soft update.

    The weights of the target_q_network are updated using the soft update rule:

                    w_target = (TAU * w) + (1 - TAU) * w_target

    where w_target are the weights of the target_q_network, TAU is the soft update
    parameter, and w are the weights of the q_network.

    Args:
        q_network (tf.keras.Sequential):
            The Q-Network.
        target_q_network (tf.keras.Sequential):
            The Target Q-Network.
    """

    for target_weights, q_net_weights in zip(
        target_q_network.weights, q_network.weights
    ):
        target_weights.assign(TAU * q_net_weights + (1.0 - TAU) * target_weights)


def plot_history(point_history, **kwargs):
    """
    Plots the total number of points received by the agent after each episode together
    with the moving average (rolling mean).

    Args:
        point_history (list):
            A list containing the total number of points the agent received after each
            episode.
        **kwargs: optional
            window_size (int):
                Size of the window used to calculate the moving average (rolling mean).
                This integer determines the fixed number of data points used for each
                window. The default window size is set to 10% of the total number of
                data points in point_history, i.e. if point_history has 200 data points
                the default window size will be 20.
            lower_limit (int):
                The lower limit of the x-axis in data coordinates. Default value is 0.
            upper_limit (int):
                The upper limit of the x-axis in data coordinates. Default value is
                len(point_history).
            plot_rolling_mean_only (bool):
                If True, only plots the moving average (rolling mean) without the point
                history. Default value is False.
            plot_data_only (bool):
                If True, only plots the point history without the moving average.
                Default value is False.
    """

    lower_limit = 0
    upper_limit = len(point_history)

    window_size = (upper_limit * 10) // 100

    plot_rolling_mean_only = False
    plot_data_only = False

    if kwargs:
        if "window_size" in kwargs:
            window_size = kwargs["window_size"]

        if "lower_limit" in kwargs:
            lower_limit = kwargs["lower_limit"]

        if "upper_limit" in kwargs:
            upper_limit = kwargs["upper_limit"]

        if "plot_rolling_mean_only" in kwargs:
            plot_rolling_mean_only = kwargs["plot_rolling_mean_only"]

        if "plot_data_only" in kwargs:
            plot_data_only = kwargs["plot_data_only"]

    points = point_history[lower_limit:upper_limit]

    # Generate x-axis for plotting.
    episode_num = [x for x in range(lower_limit, upper_limit)]

    # Use Pandas to calculate the rolling mean (moving average).
    rolling_mean = pd.DataFrame(points).rolling(window_size).mean()

    plt.figure(figsize=(10, 7), facecolor="white")

    if plot_data_only:
        plt.plot(episode_num, points, linewidth=1, color="cyan")
    elif plot_rolling_mean_only:
        plt.plot(episode_num, rolling_mean, linewidth=2, color="magenta")
    else:
        plt.plot(episode_num, points, linewidth=1, color="cyan")
        plt.plot(episode_num, rolling_mean, linewidth=2, color="magenta")

    text_color = "black"

    ax = plt.gca()
    ax.set_facecolor("black")
    plt.grid()
    plt.xlabel("Episode", color=text_color, fontsize=30)
    plt.ylabel("Total Points", color=text_color, fontsize=30)
    yNumFmt = mticker.StrMethodFormatter("{x:,}")
    ax.yaxis.set_major_formatter(yNumFmt)
    ax.tick_params(axis="x", colors=text_color)
    ax.tick_params(axis="y", colors=text_color)
    plt.show()


def display_table(current_state, action, next_state, reward, done):
    """
    Displays a table containing the current state, action, next state, reward, and done
    values from Gym's Lunar Lander environment.

    All floating point numbers in the table are displayed rounded to 3 decimal places
    and actions are displayed using their labels instead of their numerical value (i.e
    if action = 0, the action will be printed as "Do nothing" instead of "0").

    Args:
        current_state (numpy.ndarray):
            The current state vector returned by the Lunar Lander environment
            before an action is taken
        action (int):
            The action taken by the agent. In the Lunar Lander environment, actions are
            represented by integers in the closed interval [0,3] corresponding to:
                - Do nothing = 0
                - Fire right engine = 1
                - Fire main engine = 2
                - Fire left engine = 3
        next_state (numpy.ndarray):
            The state vector returned by the Lunar Lander environment after the agent
            takes an action, i.e the observation returned after running a single time
            step of the environment's dynamics using env.step(action).
        reward (numpy.float64):
            The reward returned by the Lunar Lander environment after the agent takes an
            action, i.e the reward returned after running a single time step of the
            environment's dynamics using env.step(action).
        done (bool):
            The done value returned by the Lunar Lander environment after the agent
            takes an action, i.e the done value returned after running a single time
            step of the environment's dynamics using env.step(action).

    Returns:
        table (Pandas Dataframe):
            A dataframe containing the current_state, action, next_state, reward,
            and done values. This will result in the table being displayed in the
            Jupyter Notebook.
    """

    STATE_VECTOR_COL_NAME = 'State Vector'
    DERIVED_COL_NAME = 'Derived from the State Vector (the closer to zero, the better)'

    # States
    add_derived_info = lambda state: np.hstack([
        state,
        [(state[0]**2 + state[1]**2)**.5],
        [(state[2]**2 + state[3]**2)**.5],
        [np.abs(state[4])]
    ])

    modified_current_state = add_derived_info(current_state)
    modified_next_state = add_derived_info(next_state)

    states = np.vstack([
        modified_current_state,
        modified_next_state,
        modified_next_state - modified_current_state,
    ]).T

    get_state = lambda idx, type=np.float32: dict(zip(
        ['Current State', 'Next State'],
        states[idx].astype(type)
    ))

    # Actions
    action_labels = [
        "Do nothing",
        "Fire right engine",
        "Fire main engine",
        "Fire left engine",
    ]

    display(
        pd.DataFrame({
            ('', '', ''): {'Action': action_labels[action], 'Reward': reward, 'Episode Terminated': done},
            (STATE_VECTOR_COL_NAME, 'Coordinate', 'X (Horizontal)'): get_state(0),
            (STATE_VECTOR_COL_NAME, 'Coordinate', 'Y (Vertical)'): get_state(1),
            (STATE_VECTOR_COL_NAME, 'Velocity', 'X (Horizontal)'): get_state(2),
            (STATE_VECTOR_COL_NAME, 'Velocity', 'Y (Vertical)'): get_state(3),
            (STATE_VECTOR_COL_NAME, 'Tilting', 'Angle'): get_state(4),
            (STATE_VECTOR_COL_NAME, 'Tilting', 'Angular Velocity'): get_state(5),
            (STATE_VECTOR_COL_NAME, 'Ground contact', 'Left Leg?'): get_state(6, np.bool_),
            (STATE_VECTOR_COL_NAME, 'Ground contact', 'Right Leg?'): get_state(7, np.bool_),
            (DERIVED_COL_NAME, 'Distance from landing pad', ''): get_state(8),
            (DERIVED_COL_NAME, 'Velocity', ''): get_state(9),
            (DERIVED_COL_NAME, 'Tilting Angle (absolute value)', ''): get_state(10),
        })\
            .fillna('')\
            .reindex(['Current State', 'Action', 'Next State', 'Reward', 'Episode Terminated'])\
            .style\
            .applymap(lambda x: 'background-color : grey' if x == '' else '')\
            .set_table_styles(
                [
                    {"selector": "th", "props": [("border", "1px solid grey"), ('text-align', 'center')]},
                    {"selector": "tbody td", "props": [("border", "1px solid grey"), ('text-align', 'center')]},
                ]
            )
    )


def embed_mp4(filename):
    """
    Embeds an MP4 video file in a Jupyter notebook.

    Args:
        filename (string):
            The path to the the MP4 video file that will be embedded (i.e.
            "./videos/lunar_lander.mp4").

    Returns:
        Returns a display object from the given video file. This will result in the
        video being displayed in the Jupyter Notebook.
    """

    video = open(filename, "rb").read()
    b64 = base64.b64encode(video)
    tag = """
    <video width="840" height="480" controls>
    <source src="data:video/mp4;base64,{0}" type="video/mp4">
    Your browser does not support the video tag.
    </video>""".format(
        b64.decode()
    )

    return IPython.display.HTML(tag)


def create_video(filename, env, q_network, fps=30):
    """
    Creates a video of an agent interacting with a Gym environment.

    The agent will interact with the given env environment using the q_network to map
    states to Q values and using a greedy policy to choose its actions (i.e it will
    choose the actions that yield the maximum Q values).

    The video will be saved to a file with the given filename. The video format must be
    specified in the filename by providing a file extension (.mp4, .gif, etc..). If you
    want to embed the video in a Jupyter notebook using the embed_mp4 function, then the
    video must be saved as an MP4 file.

    Args:
        filename (string):
            The path to the file to which the video will be saved. The video format will
            be selected based on the filename. Therefore, the video format must be
            specified in the filename by providing a file extension (i.e.
            "./videos/lunar_lander.mp4"). To see a list of supported formats see the
            imageio documentation: https://imageio.readthedocs.io/en/v2.8.0/formats.html
        env (Gym Environment):
            The Gym environment the agent will interact with.
        q_network (tf.keras.Sequential):
            A TensorFlow Keras Sequential model that maps states to Q values.
        fps (int):
            The number of frames per second. Specifies the frame rate of the output
            video. The default frame rate is 30 frames per second.
    """

    with imageio.get_writer(filename, fps=fps) as video:
        done = False
        state = env.reset()
        frame = env.render(mode="rgb_array")
        video.append_data(frame)
        while not done:
            state = np.expand_dims(state, axis=0)
            q_values = q_network(state)
            action = np.argmax(q_values.numpy()[0])
            state, _, done, _ = env.step(action)
            frame = env.render(mode="rgb_array")
            video.append_data(frame)
````

### *Deep Q-Learning*  

Quando tanto o espaço de estados quanto o espaço de ações são discretos, podemos estimar a função de valor de ação iterativamente utilizando a equação de Bellman:  

$$
Q_{i+1}(s,a) = R + \gamma \max_{a'}Q_i(s',a')
$$

Este método iterativo converge para a função de valor de ação ótima $ Q^*(s,a) $ conforme $ i \to \infty $. Isso significa que o agente só precisa explorar gradualmente o espaço de estados e ações e continuar atualizando a estimativa de $ Q(s,a) $ até que ela converja para $ Q^*(s,a) $.  

No entanto, quando o espaço de estados é contínuo, torna-se praticamente impossível explorar todo o espaço de estados e ações. Consequentemente, isso também torna inviável estimar gradualmente $ Q(s,a) $ até que ele atinja $ Q^*(s,a) $.  

No *Deep Q-Learning*, resolvemos esse problema utilizando uma **rede neural** para estimar a função de valor de ação, ou seja, $ Q(s,a) \approx Q^*(s,a) $. Chamamos essa rede neural de **Q-Network**, e ela pode ser treinada ajustando seus pesos a cada iteração para minimizar o erro quadrático médio na equação de Bellman.  

Infelizmente, o uso de redes neurais no aprendizado por reforço para estimar funções de valor de ação tem se mostrado altamente instável. Felizmente, existem algumas técnicas que podem ser empregadas para evitar essas instabilidades.  

Essas técnicas consistem no uso de:  
- **Rede Alvo (*Target Network*)**  
- **Reprodução de Experiência (*Experience Replay*)**  

Exploraremos essas duas técnicas nas próximas seções.

Para resolver o ambiente *Lunar Lander*, utilizaremos uma **DQN** com a seguinte arquitetura:  

1. Uma camada de entrada (`Input`) que recebe `state_size` como entrada.  
2. Uma camada **densa** (`Dense`) com **64 neurônios** e ativação **ReLU**.  
3. Outra camada **densa** (`Dense`) com **64 neurônios** e ativação **ReLU**.  
4. Uma camada de saída (`Dense`) com **`num_actions` neurônios** e ativação **linear** (pois estamos estimando valores contínuos de $ Q $).  

Ambas as redes (**Q-Network** e **Target $\hat{Q}$-Network**) devem ter a **mesma arquitetura**.  

Além disso, você deve configurar o **otimizador Adam** com uma **taxa de aprendizado (`learning rate`) igual a `ALPHA`**, que foi definida anteriormente na seção de [Hiperparâmetros].

````python
# Create the Q-Network
q_network = Sequential([
    Input(shape=state_size),
    Dense(64, activation='relu'),
    Dense(64, activation='relu'),
    Dense(num_actions)
    ])

# Create the target Q^-Network
target_q_network = Sequential([
    Input(shape=state_size),
    Dense(64, activation='relu'),
    Dense(64, activation='relu'),
    Dense(num_actions)
    ])

optimizer = Adam(learning_rate=ALPHA)
````

### *Experience Replay*  

Quando um agente interage com o ambiente, os estados, ações e recompensas que ele experimenta ocorrem de forma **sequencial**. Se o agente tentar aprender diretamente a partir dessas experiências consecutivas, pode enfrentar problemas devido às **fortes correlações** entre elas.  

Para evitar esse problema, utilizamos uma técnica chamada **Experience Replay** (*Reprodução de Experiência*), que permite gerar **experiências não correlacionadas** para treinar o agente.  

---

### **Como funciona o *Experience Replay***  

**Armazenamento das experiências**:  
   - Todas as experiências do agente (isto é, os estados, ações e recompensas que ele recebe) são armazenadas em um **buffer de memória**.  
   - Cada experiência é representada por uma tupla $ (S_t, A_t, R_t, S_{t+1}) $, onde:  
     - $ S_t $ = Estado atual  
     - $ A_t $ = Ação tomada  
     - $ R_t $ = Recompensa recebida  
     - $ S_{t+1} $ = Novo estado após a ação  

**Amostragem aleatória para aprendizado**:  
   - Em vez de aprender diretamente a partir das experiências mais recentes (o que pode levar a aprendizado instável), amostramos um **mini-lote aleatório** (*mini-batch*) de experiências do buffer de memória.  
   - Isso **quebra as correlações** entre as experiências consecutivas e melhora a **eficiência do aprendizado**.  

````python
# Store experiences as named tuples
experience = namedtuple("Experience", field_names=["state", "action", "reward", "next_state", "done"])
````

Alguns pontos a serem observados:

- A função `compute_loss` recebe um mini-lote de tuplas de experiência. Esse mini-lote é desempacotado para extrair as variáveis `states`, `actions`, `rewards`, `next_states` e `done_vals`. Você deve ter em mente que essas variáveis são *Tensors* do TensorFlow, cujo tamanho dependerá do tamanho do mini-lote. Por exemplo, se o tamanho do mini-lote for `64`, tanto `rewards` quanto `done_vals` serão Tensors do TensorFlow com `64` elementos.

- Utilizar declarações `if/else` para definir os alvos $ y $ não funcionará quando as variáveis são Tensors com muitos elementos. Contudo, note que você pode utilizar os `done_vals` para implementar o procedimento acima em uma única linha de código. Para isso, lembre-se que a variável `done` é um valor Booleano que assume `True` quando um episódio termina no passo $ j+1 $ e `False` caso contrário. Considerando que o valor Booleano `True` tem o valor numérico `1` e `False` tem o valor numérico `0`, você pode usar o fator `(1 - done_vals)` para implementar o que foi descrito em uma única linha de código. Dica: observe que `(1 - done_vals)` terá valor `0` quando `done_vals` for `True` e valor `1` quando `done_vals` for `False`.

Por fim, compute o erro calculando o Erro Quadrático Médio (*Mean-Squared Error*, MSE) entre os `y_targets` e os `q_values`. Para calcular o erro quadrático médio, utilize o pacote já importado `MSE`:

````python
def compute_loss(experiences, gamma, q_network, target_q_network):
    """
    Calculates the loss.

    Args:
      experiences: (tuple) tuple of ["state", "action", "reward", "next_state", "done"] namedtuples
      gamma: (float) The discount factor.
      q_network: (tf.keras.Sequential) Keras model for predicting the q_values
      target_q_network: (tf.keras.Sequential) Keras model for predicting the targets

    Returns:
      loss: (TensorFlow Tensor(shape=(0,), dtype=int32)) the Mean-Squared Error between
            the y targets and the Q(s,a) values.
    """

    # Unpack the mini-batch of experience tuples
    states, actions, rewards, next_states, done_vals = experiences

    # Compute max Q^(s,a)
    max_qsa = tf.reduce_max(target_q_network(next_states), axis=-1)

    # Set y = R if episode terminates, otherwise set y = R + γ max Q^(s,a).
    y_targets = rewards + gamma * max_qsa * (1 - done_vals)


    # Get the q_values and reshape to match y_targets
    q_values = q_network(states)
    q_values = tf.gather_nd(q_values, tf.stack([tf.range(q_values.shape[0]),
                                                tf.cast(actions, tf.int32)], axis=1))

    # Compute the loss
    loss = MSE(y_targets, q_values)

    return loss
````

## Atualizar os Pesos da Rede  

A função `agent_learn` atualizará os pesos das redes **Q-Network** e **Target \(\hat{Q}\)-Network** utilizando um **loop de treinamento personalizado** (*custom training loop*).  

---

### **Passos do Algoritmo**  

Como estamos usando um **loop de treinamento personalizado**, precisamos recuperar os **gradientes** utilizando uma instância do `tf.GradientTape`.  
Depois, chamamos `optimizer.apply_gradients()` para **atualizar os pesos da Q-Network**.  
Estamos utilizando o **decorador `@tf.function`** para melhorar o desempenho. Sem esse decorador, o treinamento levaria **o dobro do tempo**. Se quiser saber mais sobre como aumentar a performance com `@tf.function`, consulte a [documentação do TensorFlow](https://www.tensorflow.org/guide/function).  
 A **última linha** da função **atualiza os pesos da Target $\hat{Q}$-Network** utilizando uma **atualização suave (*soft update*)**.  

````python
@tf.function
def agent_learn(experiences, gamma):
    """
    Updates the weights of the Q networks.

    Args:
      experiences: (tuple) tuple of ["state", "action", "reward", "next_state", "done"] namedtuples
      gamma: (float) The discount factor.

    """

    # Calculate the loss
    with tf.GradientTape() as tape:
        loss = compute_loss(experiences, gamma, q_network, target_q_network)

    # Get the gradients of the loss with respect to the weights.
    gradients = tape.gradient(loss, q_network.trainable_variables)

    # Update the weights of the q_network.
    optimizer.apply_gradients(zip(gradients, q_network.trainable_variables))

    # update the weights of target q_network
    update_target_network(q_network, target_q_network)
````

 **Linha 1**: Inicializamos o `memory_buffer` com capacidade $ N = $ `MEMORY_SIZE`. Utilizamos um **`deque`** como estrutura de dados para armazenar as experiências.  

 **Linha 2**: Pulamos esta linha, pois já inicializamos a `q_network`.  

 **Linha 3**: Inicializamos a `target_q_network`, definindo seus pesos **iguais** aos da `q_network`.  

 **Linha 4**: Iniciamos o **loop externo**, onde definimos $ M = $ `num_episodes = 2000`. Esse número é razoável, pois o agente **deve ser capaz de resolver o ambiente *Lunar Lander* em menos de 2000 episódios** usando os parâmetros padrão deste notebook.  

 **Linha 5**: Utilizamos o método `.reset()` para reinicializar o ambiente no estado inicial e obter esse **estado inicial**.  

 **Linha 6**: Iniciamos o **loop interno**, definindo $ T = $ `max_num_timesteps = 1000`. Isso significa que o episódio **terminará automaticamente** caso ultrapasse 1000 passos de tempo.  

 **Linha 7**: O agente observa o `state` atual e escolhe uma `action` usando a **política ε-greedy**.  
   - O valor inicial de $ \epsilon = $ `1` cria uma **política aleatória**, ou seja, o agente toma **ações aleatórias no início do treinamento**.  
   - À medida que o treinamento progride, o valor de $ \epsilon $ **diminui gradualmente** até atingir um valor mínimo, conforme a taxa de **decaimento de $ \epsilon $**.  
   - Definimos o mínimo de $ \epsilon = 0.01 $ para manter **um pequeno nível de exploração** durante o treinamento.  
   - Se quiser entender como isso é implementado, veja a função `get_action`.  

 **Linha 8**: Utilizamos `.step()` para executar a `action` no ambiente e obter os valores de `reward` e `next_state`.  

 **Linha 9**: Armazenamos a **tupla de experiência** `(state, action, reward, next_state, done)` no `memory_buffer`.  
   - Também armazenamos a variável `done` para saber quando um episódio termina.  
   - Isso nos permitiu definir os alvos $ y $.  

 **Linha 10**: Verificamos se as **condições para atualização** foram atendidas.  
    - Utilizamos a função `utils.check_update_conditions`, que verifica se **$ C = $ `NUM_STEPS_FOR_UPDATE = 4`** passos de tempo ocorreram e se o `memory_buffer` tem **experiências suficientes para preencher um mini-lote**.  
    - Por exemplo, se o tamanho do mini-lote for `64`, o `memory_buffer` precisa ter **pelo menos 64 tuplas de experiência** para que a atualização ocorra.  
    - Se as condições forem atendidas, `check_update_conditions` retorna `True`, caso contrário, retorna `False`.  

 **Linhas 11 - 14**: Se `update = True`, realizamos a **atualização do aprendizado**:  
    - **Amostramos um mini-lote aleatório** de experiências do `memory_buffer`.  
    - **Calculamos os alvos $ y $** para a função de perda.  
    - **Realizamos descida do gradiente** para atualizar os pesos da **Q-Network**.  
    - **Atualizamos os pesos da *Target Network*** utilizando o método de **atualização suave** (*soft update*).  
    - Usamos a função `agent_learn` para realizar essas etapas.  

 **Linha 15**: No final de cada iteração do **loop interno**:  
    - Definimos `next_state` como o **novo `state`**, permitindo que o loop continue a partir desse novo estado.  
    - Verificamos se o episódio atingiu um **estado terminal** (`done = True`). Se sim, saímos do **loop interno**.  

 **Linha 16**: No final de cada iteração do **loop externo**:  
    - Atualizamos o valor de $ \epsilon $.  
    - Verificamos se o ambiente foi **resolvido**.  
      - O ambiente é considerado **resolvido** se o agente atingir uma média de **200 pontos** nos últimos **100 episódios**.  
    - Se o ambiente ainda **não** foi resolvido, continuamos o **loop externo**, iniciando um **novo episódio**.  

---

### **Medição de Desempenho e Tempo**  

Além das operações acima, incluímos algumas variáveis extras para acompanhar:  

- O **número total de pontos** que o agente recebe em cada episódio.  
- Isso nos ajuda a verificar se o agente **resolveu o ambiente** e também a visualizar **como o desempenho do agente evolui** ao longo do treinamento.  
- Utilizamos o módulo `time` para medir **quanto tempo o treinamento leva**.  

````python
start = time.time()

num_episodes = 2000
max_num_timesteps = 1000

total_point_history = []

num_p_av = 100    # number of total points to use for averaging
epsilon = 1.0     # initial ε value for ε-greedy policy

# Create a memory buffer D with capacity N
memory_buffer = deque(maxlen=MEMORY_SIZE)

# Set the target network weights equal to the Q-Network weights
target_q_network.set_weights(q_network.get_weights())

for i in range(num_episodes):

    # Reset the environment to the initial state and get the initial state
    state = env.reset()
    total_points = 0

    for t in range(max_num_timesteps):

        # From the current state S choose an action A using an ε-greedy policy
        state_qn = np.expand_dims(state, axis=0)  # state needs to be the right shape for the q_network
        q_values = q_network(state_qn)
        action = get_action(q_values, epsilon)

        # Take action A and receive reward R and the next state S'
        next_state, reward, done, _ = env.step(action)

        # Store experience tuple (S,A,R,S') in the memory buffer.
        # We store the done variable as well for convenience.
        memory_buffer.append(experience(state, action, reward, next_state, done))

        # Only update the network every NUM_STEPS_FOR_UPDATE time steps.
        update = check_update_conditions(t, NUM_STEPS_FOR_UPDATE, memory_buffer)

        if update:
            # Sample random mini-batch of experience tuples (S,A,R,S') from D
            experiences = get_experiences(memory_buffer)

            # Set the y targets, perform a gradient descent step,
            # and update the network weights.
            agent_learn(experiences, GAMMA)

        state = next_state.copy()
        total_points += reward

        if done:
            break

    total_point_history.append(total_points)
    av_latest_points = np.mean(total_point_history[-num_p_av:])

    # Update the ε value
    epsilon = get_new_eps(epsilon)

    print(f"\rEpisode {i+1} | Total point average of the last {num_p_av} episodes: {av_latest_points:.2f}", end="")

    if (i+1) % num_p_av == 0:
        print(f"\rEpisode {i+1} | Total point average of the last {num_p_av} episodes: {av_latest_points:.2f}")

    # We will consider that the environment is solved if we get an
    # average of 200 points in the last 100 episodes.
    if av_latest_points >= 200.0:
        print(f"\n\nEnvironment solved in {i+1} episodes!")
        q_network.save('lunar_lander_model.h5')
        break

tot_time = time.time() - start

print(f"\nTotal Runtime: {tot_time:.2f} s ({(tot_time/60):.2f} min)")
````
