````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
````

#### **Aprendizado Supervisionado**:
O aprendizado supervisionado é um tipo de aprendizado de máquina onde o modelo é treinado usando um conjunto de dados rotulados. Isso significa que para cada entrada, há uma saída conhecida, permitindo que o modelo aprenda a mapear entradas para saídas.

#### **Regressão Linear**:
A regressão linear é um dos modelos mais simples e amplamente usados no aprendizado supervisionado. O objetivo é ajustar uma linha reta aos dados que melhor explica a relação entre a variável independente (entrada) e a variável dependente (saída).

#### **Exemplo Prático**:
Vamos considerar um exemplo onde queremos prever o preço de uma casa com base no tamanho da casa. Temos um conjunto de dados que contém tamanhos de casas e seus respectivos preços.

#### **Visualização dos Dados**:
Os dados são geralmente visualizados em um gráfico de dispersão, onde cada ponto representa uma casa com seu tamanho (eixo x) e preço (eixo y).

#### **Modelo de Regressão Linear**:
O modelo de regressão linear ajusta uma linha reta aos dados. A equação da linha é geralmente dada por:
$$ y = mx + c $$
onde $ y $ é o preço previsto, $ m $ é a inclinação da linha, $ x $ é o tamanho da casa e $ c $ é o intercepto.

#### **Notação**:
- **x**: Variável de entrada (tamanho da casa).
- **y**: Variável de saída (preço da casa).
- **(x, y)**: Um exemplo de treinamento.
- **m**: Número total de exemplos de treinamento.

#### **Objetivo**:
Dado um novo valor de $ x $ (tamanho de uma casa), queremos prever o valor correspondente de $ y $ (preço da casa).

#### **Implementação em Python**:
Vamos usar Python para implementar este modelo, ajustar a linha de regressão e prever o preço de uma casa com 1250 pés quadrados.

````python
# Dados de exemplo: tamanho da casa (em pés quadrados) e preços (em milhares de dólares)
data = {
    'Tamanho': [1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200, 2300, 2400, 2500],
    'Preço': [300, 320, 340, 360, 380, 400, 420, 440, 460, 480, 500]
}

# Criar um DataFrame
df = pd.DataFrame(data)

# Visualizar os dados
plt.scatter(df['Tamanho'], df['Preço'], color='blue', marker='x')
plt.xlabel('Tamanho da Casa (pés quadrados)')
plt.ylabel('Preço da Casa (milhares de dólares)')
plt.title('Tamanho da Casa vs. Preço da Casa')
plt.show()

# Criar e treinar o modelo de regressão linear
X = df[['Tamanho']]  # Variável independente (entrada)
y = df['Preço']  # Variável dependente (saída)

model = LinearRegression()
model.fit(X, y)

# Prever o preço de uma casa com 1250 pés quadrados
tamanho_casa = pd.DataFrame({'Tamanho': [1250]})
preco_previsto = model.predict(tamanho_casa)
print(f'O preço previsto para uma casa com 1250 pés quadrados é: ${preco_previsto[0]*1000:.2f}')

# Visualizar a linha de regressão
plt.scatter(df['Tamanho'], df['Preço'], color='blue', marker='x')
plt.plot(df['Tamanho'], model.predict(X), color='red')
plt.xlabel('Tamanho da Casa (pés quadrados)')
plt.ylabel('Preço da Casa (milhares de dólares)')
plt.title('Tamanho da Casa vs. Preço da Casa')
plt.show()
````

#### **Modelo de Regressão Linear**:
Na regressão linear, a função $ f $ que estamos tentando aprender é uma linha reta. A fórmula para esta linha é:
$$ f(x) = wx + b $$
onde:
- $ x $ é a entrada (por exemplo, o tamanho da casa).
- $ w $ é o coeficiente angular (inclinação da linha).
- $ b $ é o intercepto (onde a linha cruza o eixo y).
- $ y $ é a saída (por exemplo, o preço da casa).
- $ \hat{y} $ é a predição ou estimativa de $ y $.

#### **Função de Custo**:
Para ajustar a melhor linha reta aos dados, precisamos medir o quão bem uma linha específica se ajusta aos nossos dados de treinamento. Fazemos isso usando uma função de custo. A função de custo é uma métrica que quantifica o erro entre as previsões do modelo $ \hat{y} $ e os valores reais $ y $.

Uma função de custo comum para regressão linear é a **Erro Quadrático Médio (Mean Squared Error - MSE)**:
$$ J(w, b) = \frac{1}{2m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})^2 $$
onde:
- $ m $ é o número de exemplos de treinamento.
- $ \hat{y}^{(i)} $ é a predição do modelo para o exemplo $ i $.
- $ y^{(i)} $ é o valor real para o exemplo $ i $.

#### **Minimização da Função de Custo**:
O objetivo do aprendizado é encontrar os valores de $ w $ e $ b $ que minimizam a função de custo $ J(w, b) $. Isso é feito utilizando algoritmos de otimização, como o gradiente descendente.

#### **Gradiente Descendente**:
O gradiente descendente é um método iterativo para encontrar os mínimos de uma função. A ideia é começar com valores iniciais de $ w $ e $ b $ e ajustar esses valores iterativamente na direção oposta ao gradiente da função de custo.

As atualizações para $ w $ e $ b $ são dadas por:
$$ w := w - \alpha \frac{\partial J(w, b)}{\partial w} $$
$$ b := b - \alpha \frac{\partial J(w, b)}{\partial b} $$
onde:
- $ \alpha $ é a taxa de aprendizado, um hiperparâmetro que controla o tamanho do passo de cada iteração.

````python
# Dados de exemplo: tamanho da casa (em pés quadrados) e preços (em milhares de dólares)
data = {
    'Tamanho': [1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200, 2300, 2400, 2500],
    'Preço': [300, 320, 340, 360, 380, 400, 420, 440, 460, 480, 500]
}

# Converter dados para numpy arrays
X = np.array(data['Tamanho'])
y = np.array(data['Preço'])

# Inicializar parâmetros
w = 0
b = 0
alpha = 0.0000001 # com uma escala de dados tão grande, precisamos de um alpha muito pequeno pois assim 
                  # garantimos que as mudanças nos pesos e bias sejam pequenas, ou podemos normalizar os dados
num_iters = 100

# Função de custo
def compute_cost(X, y, w, b):
    m = len(y)
    cost = (1/(2*m)) * np.sum((w*X + b - y)**2)
    return cost

# Gradiente descendente
def gradient_descent(X, y, w, b, alpha, num_iters):
    m = len(y)
    for i in range(num_iters):
        # Atualizar parâmetros
        dw = (1/m) * np.sum((w*X + b - y) * X)
        db = (1/m) * np.sum(w*X + b - y)
        w = w - alpha * dw
        b = b - alpha * db
        if i % 10 == 0:
            print(f"Iteração {i}: Custo {compute_cost(X, y, w, b):.5f}")
    return w, b

# Treinar o modelo
w, b = gradient_descent(X, y, w, b, alpha, num_iters)
print(f"Parâmetros treinados: w = {w:.5f}, b = {b:.5f}")

# Prever o preço de uma casa com 1250 pés quadrados
tamanho_casa = 1250
preco_previsto = w * tamanho_casa + b
print(f"O preço previsto para uma casa com 1250 pés quadrados é: ${preco_previsto * 1000:.2f}")

# Visualizar a linha de regressão
plt.scatter(X, y, color='blue', marker='x', label='Dados')
plt.plot(X, w*X + b, color='red', label='Linha de Regressão')
plt.xlabel('Tamanho da Casa (pés quadrados)')
plt.ylabel('Preço da Casa (milhares de dólares)')
plt.title('Tamanho da Casa vs. Preço da Casa')
plt.legend()
plt.show()
````

#### **Função de Custo:**

A função de custo é uma ferramenta essencial em aprendizado supervisionado para medir a discrepância entre as previsões do modelo e os valores reais dos dados de treinamento. O objetivo é minimizar essa função de custo para obter os melhores parâmetros do modelo.

#### **Visualização da Função de Custo:**

Para compreender melhor a função de custo, vamos analisar como diferentes valores do parâmetro $ w $ afetam a função de custo $ J(w) $. Como mencionado, quando ajustamos $ w $ em nosso modelo de regressão linear simples $ f_w(x) = w \cdot x $, podemos calcular a função de custo $ J(w) $ para diferentes valores de $ w $.

#### **Exemplo com $ b = 0 $:**

Vamos considerar um exemplo simplificado onde $ b = 0 $, então a função do modelo é $ f_w(x) = w \cdot x $. Neste caso, a função de custo é:

$$ J(w) = \frac{1}{2m} \sum_{i=1}^{m} (w \cdot x^{(i)} - y^{(i)})^2 $$

Aqui, $ m $ é o número de exemplos de treinamento, $ x^{(i)} $ é a entrada do i-ésimo exemplo e $ y^{(i)} $ é o valor real de saída.

### Implementação em Python:

Vamos implementar a função de custo e visualizar como ela muda para diferentes valores de $ w $.

````python
# Inicializar parâmetros
w_vals = np.linspace(-1, 1, 100) # Valores de w
J_vals = []

# Função de custo para um dado valor de w
def compute_cost_single_w(X, y, w):
    m = len(y)
    cost = (1 / (2 * m)) * np.sum((w * X - y) ** 2)
    return cost

# Calcular a função de custo para cada valor de w
for w in w_vals:
    J_vals.append(compute_cost_single_w(X, y, w))

# Plotar a função de custo
plt.plot(w_vals, J_vals, label='J(w)')
plt.xlabel('w')
plt.ylabel('J(w)')
plt.title('Função de Custo J(w) vs. Parâmetro w')
plt.legend()
plt.show()
````

### Compreensão Intuitiva:

A curva da função de custo \( J(w) \) geralmente tem a forma de uma parábola. O objetivo do treinamento do modelo é encontrar o valor de \( w \) que minimiza \( J(w) \). No gráfico, esse valor corresponde ao ponto mais baixo da parábola.

#### **Extensão para \( w \) e \( b \):**

Na prática, usamos tanto \( w \) quanto \( b \) no modelo de regressão linear \( f_w, b(x) = w \cdot x + b \). A função de custo se torna uma superfície em 3D que depende de ambos os parâmetros. A busca pelos melhores valores de \( w \) e \( b \) envolve encontrar o ponto na superfície onde \( J(w, b) \) é mínimo.

### Visualização 3D da Função de Custo:

Podemos visualizar a função de custo em 3D para obter uma compreensão melhor de como \( w \) e \( b \) afetam \( J(w, b) \).



````python
# Faixa de valores para w e b
w_vals = np.linspace(-1, 2, 100)
b_vals = np.linspace(-1, 2, 100)
W, B = np.meshgrid(w_vals, b_vals)
J_vals = np.zeros_like(W)

# Calcular a função de custo para cada par (w, b)
for i in range(len(w_vals)):
    for j in range(len(b_vals)):
        J_vals[i, j] = compute_cost(X, y, W[i, j], B[i, j])

# Plotar a superfície da função de custo
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, J_vals, cmap='viridis')
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('J(w, b)')
ax.set_title('Superfície da Função de Custo')
plt.show()
````

#### **Introdução ao Gradiente Descendente**

Gradiente Descendente é um dos algoritmos fundamentais para otimização em aprendizado de máquina. Ele é utilizado para minimizar uma função de custo, ajustando iterativamente os parâmetros do modelo. Na regressão linear, esses parâmetros são $ w $ (peso) e $ b $ (interceptação), que determinam a linha que melhor se ajusta ao conjunto de dados.

#### **Funcionamento do Gradiente Descendente**

O objetivo do Gradiente Descendente é encontrar os valores de $ w $ e $ b $ que minimizam a função de custo $ J(w, b) $. O processo é iterativo e funciona da seguinte maneira:

1. **Inicialização**: Comece com valores iniciais para $ w $ e $ b $. Esses podem ser zeros, valores aleatórios ou baseados em alguma heurística.

2. **Cálculo do Gradiente**: Em cada etapa, calcule o gradiente da função de custo em relação a cada parâmetro. O gradiente é um vetor que aponta na direção do maior aumento da função. Portanto, para minimizar a função, precisamos mover na direção oposta ao gradiente.

3. **Atualização dos Parâmetros**: Atualize os parâmetros $ w $ e $ b $ subtraindo um múltiplo do gradiente. Esse múltiplo é determinado pela taxa de aprendizado $ \alpha $, um hiperparâmetro que define o tamanho do passo em cada iteração:
   $$
   w := w - \alpha \frac{\partial J}{\partial w}, \quad b := b - \alpha \frac{\partial J}{\partial b}
   $$

4. **Convergência**: Repita os passos 2 e 3 até que os parâmetros $ w $ e $ b $ converjam para valores que minimizem $ J $. A convergência é geralmente definida por uma pequena mudança nos valores de $ w $ e $ b $ entre iterações sucessivas ou até que a mudança no valor de $ J $ seja insignificante.

#### **Visualização e Intuição**

Para entender intuitivamente como o gradiente descendente funciona, você pode visualizar o processo em um gráfico 3D ou um gráfico de contorno da função de custo. Conforme os parâmetros $ w $ e $ b $ são ajustados, o ponto representando $ (w, b, J(w, b)) $ se move na superfície da função de custo em direção ao ponto mais baixo, que é o mínimo global.

````python
# Dados hipotéticos
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Parâmetros iniciais
w = 0
b = 0

# Taxa de aprendizado
alpha = 0.01

# Número de iterações
iterations = 25

# Armazenar histórico para plotagem
w_history = []
b_history = []
cost_history = []

# Função para calcular o custo
def compute_cost(X, y, w, b):
    m = len(y)
    predictions = w * X + b
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return cost

# Gradiente Descendente
for i in range(iterations):
    predictions = w * X + b
    error = predictions - y
    w_grad = (1 / len(X)) * np.sum(error * X)
    b_grad = (1 / len(X)) * np.sum(error)
    
    w -= alpha * w_grad
    b -= alpha * b_grad
    
    cost = compute_cost(X, y, w, b)
    
    w_history.append(w)
    b_history.append(b)
    cost_history.append(cost)

# Plotagem dos dados e da linha de regressão em cada iteração
plt.figure(figsize=(16, 7))

# Plotagem dos dados
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue', label='Dados')

# Linhas de regressão
colors = plt.cm.winter(np.linspace(0, 1, iterations))
for i in range(iterations):
    plt.plot(X, w_history[i] * X + b_history[i], color=colors[i], label=f'Iteração {i+1}')

plt.title('Regressão Linear ao Longo das Iterações')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

# Plotagem do contorno da função de custo
plt.subplot(1, 2, 2)

# Valores para W e B
W, B = np.meshgrid(np.linspace(-0.5, 2, 100), np.linspace(-1, 3, 100))
costs = np.zeros_like(W)

# Cálculo dos custos para cada par (w, b)
for i in range(W.shape[0]):
    for j in range(W.shape[1]):
        costs[i, j] = compute_cost(X, y, W[i, j], B[i, j])

# Contorno
cp = plt.contour(W, B, costs, levels=np.logspace(-2, 3, 20), cmap='viridis')
plt.clabel(cp, inline=True, fontsize=10)
plt.scatter(w_history, b_history, c='red', s=100)  # Pontos das iterações
plt.title('Contorno da Função de Custo')
plt.xlabel('w')
plt.ylabel('b')

plt.tight_layout()
plt.show()
````

### Gradiente Descendente: Uma Visão Detalhada

O Gradiente Descendente é uma técnica de otimização fundamental no campo do aprendizado de máquina, empregada para minimizar uma função de custo, que é uma medida de quão "errado" um modelo de aprendizado de máquina está em termos de sua capacidade de estimar a relação subjacente entre as variáveis independentes e dependentes nos dados.

#### **Princípios Básicos do Gradiente Descendente**

O processo básico do gradiente descendente envolve a atualização repetida dos parâmetros do modelo para reduzir a função de custo. A ideia é mover os parâmetros na direção que resulta na maior diminuição possível da função de custo. Isso é feito calculando o gradiente (ou a derivada) da função de custo em relação aos parâmetros do modelo. O gradiente é simplesmente a inclinação da função de custo em relação a cada parâmetro.

#### **Funcionamento do Gradiente Descendente**

1. **Inicialização**: O gradiente descendente começa com uma escolha inicial para os parâmetros do modelo. Esta escolha inicial pode ser aleatória ou baseada em alguma heurística.

2. **Cálculo do Gradiente**: O algoritmo calcula o gradiente da função de custo, que indica a direção do maior aumento da função. Para minimizar a função, os parâmetros devem ser atualizados na direção oposta ao gradiente.

3. **Atualização dos Parâmetros**: Os parâmetros são então ajustados subtraindo um produto do gradiente pela taxa de aprendizado, um hiperparâmetro que define o tamanho do passo da atualização. Matematicamente, isto é expresso como:
   $$
   w := w - \alpha \frac{\partial J}{\partial w}
   $$
   $$
   b := b - \alpha \frac{\partial J}{\partial b}
   $$
   Onde $ \alpha $ é a taxa de aprendizado e $ \frac{\partial J}{\partial w} $, $ \frac{\partial J}{\partial b} $ são os gradientes da função de custo em relação aos parâmetros $ w $ e $ b $.

4. **Iteração**: Este processo é repetido (iterado) muitas vezes até que a função de custo cesse de diminuir significativamente, indicando que os parâmetros encontraram os valores que minimizam a função de custo, ou até que um número máximo de iterações seja alcançado.

#### **Propriedades do Gradiente Descendente**

- **Convergência**: Em teoria, o gradiente descendente convergirá para o mínimo global se a função de custo for convexa. No entanto, para funções de custo não-convexas, como as encontradas em redes neurais, o gradiente descendente pode encontrar mínimos locais ou pontos de sela.
- **Taxa de Aprendizado**: A escolha da taxa de aprendizado é crucial; uma taxa muito alta pode fazer com que o algoritmo diverja, enquanto uma taxa muito baixa pode resultar em um processo de convergência muito lento.

#### **Aplicações e Generalizações**

O gradiente descendente não se limita à regressão linear; é aplicável a qualquer modelo de aprendizado de máquina que utilize uma função de custo diferenciável. Isso inclui redes neurais profundas, máquinas de vetores de suporte com kernels diferenciáveis, e muitos outros. Além disso, existem várias variantes do gradiente descendente que tentam melhorar sua eficiência e taxa de convergência, como o Gradiente Descendente Estocástico (SGD), Momentum, AdaGrad, RMSprop, e Adam.

#### **Desafios**

Apesar de sua utilidade, o gradiente descendente vem com desafios, principalmente relacionados à escolha de hiperparâmetros como a taxa de aprendizado e o número de iterações. Além disso, em funções de custo complexas e de alta dimensão, encontrar o mínimo global pode ser computacionalmente desafiador e exigir ajustes cuidadosos dos parâmetros iniciais e da taxa de aprendizado.

#### **Taxa de Aprendizado (Alpha)**

A taxa de aprendizado, $ \alpha $, é um dos hiperparâmetros mais críticos no gradiente descendente. Ela determina o tamanho do passo que o algoritmo dá em direção ao mínimo da função de custo a cada atualização dos parâmetros $ w $ e $ b $. A escolha adequada de $ \alpha $ pode ter um grande impacto na eficiência e sucesso do treinamento do modelo.

#### **Influência da Taxa de Aprendizado**

1. **Taxa de Aprendizado Grande Demais**: 
   - Se $ \alpha $ for muito grande, o gradiente descendente pode "saltar" sobre o mínimo, resultando em valores da função de custo que oscilam ou até aumentam a cada iteração. Em casos extremos, o algoritmo pode divergir completamente, falhando em encontrar uma solução adequada.

2. **Taxa de Aprendizado Muito Pequena**:
   - Por outro lado, se $ \alpha $ for muito pequeno, o algoritmo de gradiente descendente fará progressos muito lentos em direção ao mínimo, o que pode levar a um longo tempo de treinamento. Além disso, há um risco aumentado do algoritmo ficar preso em mínimos locais ou pontos de sela se a superfície da função de custo for complexa.

#### **Escolhendo a Taxa de Aprendizado**

A escolha da taxa de aprendizado geralmente requer experimentação e ajuste fino. Algumas estratégias comuns incluem:

- **Teste Empírico**: Executar o algoritmo várias vezes com diferentes valores de $ \alpha $ e escolher o valor que minimiza a função de custo mais rapidamente sem causar divergência.
- **Decaimento da Taxa de Aprendizado**: Reduzir $ \alpha $ gradualmente durante o treinamento pode ajudar a fazer ajustes finos à medida que o algoritmo se aproxima do mínimo.

#### **Considerações Adicionais**

- **Inicialização dos Parâmetros**: Além de escolher um bom $ \alpha $, iniciar os parâmetros $ w $ e $ b $ de forma adequada também é importante. Inicializações ruins podem levar a convergências lentas ou a mínimos locais subótimos.
- **Adaptação da Taxa de Aprendizado**: Algoritmos modernos de otimização, como Adam ou RMSprop, ajustam automaticamente a taxa de aprendizado durante o treinamento, o que pode ajudar a alcançar melhores resultados mais rapidamente e com menos esforço na sintonização de hiperparâmetros.

````python
# Função custo simples: J(w) = (w - 3)^2
def compute_cost(w):
    return (w - 3)**2

# Gradiente da função custo: dJ/dw = 2*(w - 3)
def compute_gradient(w):
    return 2 * (w - 3)

# Valores iniciais
w_start = -4
iterations = 15

# Taxas de aprendizado para teste
learning_rates = [0.01, 0.1, 0.5]

# Configurações do plot
plt.figure(figsize=(12, 8))

for alpha in learning_rates:
    w = w_start
    cost_history = []
    w_history = []
    
    for i in range(iterations):
        cost = compute_cost(w)
        gradient = compute_gradient(w)
        w = w - alpha * gradient  # Atualização do parâmetro
        cost_history.append(cost)
        w_history.append(w)
    
    # Plotagem dos resultados
    plt.plot(cost_history, label=f'α = {alpha}')

plt.title('Influência da Taxa de Aprendizado no Gradiente Descendente')
plt.xlabel('Iterações')
plt.ylabel('Custo')
plt.legend()
plt.show()
````

### Expansão da Regressão Linear para Múltiplas Variáveis

#### **Introdução à Regressão Linear Múltipla**

A regressão linear simples, que utiliza apenas uma variável (ou característica) para prever uma resposta, é uma ferramenta útil, mas muitas vezes as relações do mundo real são mais complexas e dependem de múltiplas variáveis. Por exemplo, o preço de uma casa pode depender não só do tamanho, mas também do número de quartos, número de banheiros, idade da casa, localização, entre outros fatores. A regressão linear múltipla permite incorporar essas múltiplas variáveis para fornecer uma previsão mais precisa e robusta.

#### **Notação e Modelo para Regressão Linear Múltipla**

- **Variáveis e Notações**:
  - **$ X_j $**: Representa as diferentes características (features). Por exemplo, $ X_1 $ pode ser o tamanho da casa, $ X_2 $ o número de quartos, etc.
  - **$ n $**: Denota o número total de características usadas no modelo. No exemplo dado, $ n = 4 $ (tamanho, quartos, andares, idade).
  - **$ X^{(i)} $**: Representa o vetor de características do $ i $-ésimo exemplo de treinamento.
  - **$ X_j^{(i)} $**: É o valor da $ j $-ésima característica no $ i $-ésimo exemplo de treinamento.

- **Modelo**:
  - O modelo para a regressão linear múltipla pode ser expresso como:
    $$
    f_{wb}(X) = w_1X_1 + w_2X_2 + \ldots + w_nX_n + b
    $$
  - Aqui, $ w_1, w_2, \ldots, w_n $ são os pesos atribuídos a cada característica, e $ b $ é o termo de interceptação.

#### **Interpretação dos Coeficientes**

Cada peso $ w_j $ no modelo indica o efeito de uma unidade de aumento na característica $ X_j $ sobre a variável resposta $ y $ (preço da casa, neste caso), mantendo todas as outras características constantes. Por exemplo, se $ w_2 = 4 $, isso sugere que cada adição de um quarto aumenta o preço da casa em 4 mil dólares, assumindo que todos os outros fatores permaneçam constantes.

#### **Vantagens da Regressão Linear Múltipla**

- **Maior Precisão**: Ao considerar múltiplas variáveis, o modelo pode capturar mais complexidade e nuances dos dados, o que geralmente leva a previsões mais precisas.
- **Análise mais Rica**: Permite avaliar o impacto de diferentes variáveis de forma isolada e em combinação, proporcionando insights mais profundos sobre os fatores que influenciam a variável de interesse.

#### **Desafios**

- **Colinearidade**: Se duas ou mais variáveis são altamente correlacionadas, isso pode causar problemas na estimação dos coeficientes, tornando-os instáveis e difíceis de interpretar.
- **Sobredimensionamento**: Incluir muitas variáveis pode levar a um modelo "sobrecarregado" que se ajusta demais aos dados de treinamento e não generaliza bem para novos dados.

#### **Implementação e Otimização**

Para implementar e otimizar eficientemente um modelo de regressão linear múltipla, especialmente quando o número de características é grande, técnicas de vetorização são essenciais. Vetorização permite expressar operações matemáticas sem usar loops explícitos, o que não só simplifica o código, mas também melhora significativamente a eficiência computacional.

### Vectorização em Aprendizado de Máquina

#### **Introdução à Vectorização**

A vectorização é uma técnica poderosa em programação que envolve a execução de operações em dados sem o uso explícito de loops. No contexto do aprendizado de máquina, a vectorização não apenas simplifica o código, tornando-o mais limpo e mais curto, mas também otimiza a execução, aproveitando as bibliotecas de álgebra linear numérica e o hardware especializado, como GPUs.

#### **Vantagens da Vectorização**

1. **Eficiência de Tempo**: O código vectorizado reduz significativamente o tempo de execução ao processar operações em grandes conjuntos de dados. Isso é crucial no aprendizado de máquina, onde se trabalha frequentemente com grandes volumes de dados.
   
2. **Simplificação de Código**: A vectorização pode reduzir múltiplas linhas de código a uma única linha, minimizando a complexidade e melhorando a legibilidade.

3. **Uso de Hardware Especializado**: A vectorização é especialmente projetada para tirar proveito de arquiteturas de hardware modernas, como CPUs multicore e GPUs, que são capazes de realizar operações matemáticas paralelas de forma eficiente.

#### **Exemplo Prático de Vectorização**

Suponhamos que temos um modelo de regressão linear com múltiplas variáveis, onde queremos calcular a saída predita usando a equação:
$$ f(x) = w_1x_1 + w_2x_2 + \dots + w_nx_n + b $$

**Sem Vectorização:**
- Tradicionalmente, isso poderia ser implementado com um loop que itera sobre cada elemento dos vetores $ w $ e $ x $, multiplicando-os e somando ao resultado final. Este método é não só tedioso em termos de escrita de código, mas também ineficiente em termos computacionais.

**Com Vectorização:**
- Usando uma biblioteca como NumPy, a mesma operação pode ser implementada de forma muito mais eficiente e concisa:
```python
import numpy as np

# Definindo os vetores w e x, e o escalar b
w = np.array([w1, w2, ..., wn])
x = np.array([x1, x2, ..., xn])
b = 0.0  # exemplo de inicialização

# Calculando a previsão
f = np.dot(w, x) + b
```
A função `np.dot()` realiza um produto escalar entre os vetores $ w $ e $ x $, uma operação que é altamente otimizada para executar em hardware paralelo.

#### **Por Trás das Cenas: O Que Acontece com a Vectorização**

Quando código vectorizado é executado:
- **Paralelismo**: As operações são desmembradas e executadas em paralelo, aproveitando múltiplos núcleos de processamento disponíveis em CPUs modernas e GPUs.
- **Eficiência**: As bibliotecas como NumPy são escritas em linguagens de baixo nível como C e Fortran, que interagem diretamente com o hardware, proporcionando uma execução mais rápida do que seria possível em Python puro.

### Implementando Gradiente Descendente para Regressão Linear Múltipla com Vectorização

#### **Revisão da Regressão Linear Múltipla**

Na regressão linear múltipla, em vez de usar apenas uma variável preditora, utilizamos múltiplas variáveis para prever a resposta. Isso enriquece o modelo permitindo que ele capture mais complexidades nos dados.

#### **Modelagem com Notação Vetorial**

Ao lidar com múltiplas variáveis, é eficiente organizar os coeficientes e características em vetores:
- **Vetor $ w $**: Contém os pesos para cada característica. Por exemplo, $ w = [w_1, w_2, \dots, w_n] $.
- **Vetor $ x $**: Representa as características de uma observação. Por exemplo, $ x = [x_1, x_2, \dots, x_n] $.
- O modelo de predição pode ser expresso como:
  $$
  f_{w, b}(x) = w \cdot x + b
  $$
  onde $ \cdot $ denota o produto escalar entre os vetores $ w $ e $ x $, e $ b $ é o viés ou interceptação.

#### **Função de Custo**

A função de custo $ J(w, b) $ mede o quão bem o modelo com parâmetros $ w $ e $ b $ se ajusta aos dados. Comumente, usamos o erro quadrático médio (MSE) como a função de custo em regressão linear.

#### **Gradiente Descendente com Múltiplas Variáveis**

O processo para atualizar os parâmetros é similar ao da regressão linear simples, mas adaptado para lidar com múltiplos parâmetros:
- Cada peso $ w_j $ é atualizado subtraindo um pequeno valor proporcional ao gradiente da função de custo em relação a esse peso:
  $$
  w_j := w_j - \alpha \left( \frac{\partial J}{\partial w_j} \right)
  $$
  onde $ \alpha $ é a taxa de aprendizado.
- O gradiente de $ J $ em relação a $ w_j $ é calculado considerando o erro entre as previsões do modelo e os valores verdadeiros, multiplicado pelas respectivas características.

#### **Vantagens da Vectorização na Implementação**

A vectorização torna a implementação não apenas mais rápida, mas também mais limpa e menos propensa a erros. Em Python, usando a biblioteca NumPy, a atualização dos parâmetros pode ser realizada de maneira eficiente em uma linha:
```python
w = w - alpha * np.dot(X.T, (np.dot(X, w) + b - y))
```
Esta linha de código utiliza operações vetoriais para calcular os gradientes e atualizar os parâmetros, aproveitando a capacidade de processamento paralelo dos computadores modernos.

### Escala de Recursos e Gradiente Descendente

#### **Importância da Escala de Recursos**

A escala dos recursos (feature scaling) é uma técnica crucial para melhorar a eficiência do gradiente descendente em modelos de regressão linear com múltiplas variáveis. Isso ocorre porque diferentes características geralmente têm faixas de valores muito diferentes, o que pode afetar significativamente a performance do gradiente descendente.

#### **Problemas Sem Escala Adequada**

Quando as características têm faixas de valores muito diferentes:
1. **Velocidade de Convergência**: O gradiente descendente pode convergir muito lentamente. Isso é visualizado por contornos de função de custo que são alongados, onde mudanças em uma direção têm muito mais impacto no custo do que mudanças em outra.
2. **Oscilações**: Sem uma escala adequada, o gradiente descendente pode oscilar ineficientemente ao longo das contornos, demorando mais para alcançar o mínimo global.

#### **Normalização e Padronização**

Duas técnicas comuns para escalonamento são:
- **Normalização**: Ajusta os dados para que seus valores fiquem entre 0 e 1.
- **Padronização (Standardization)**: Transforma os dados para que tenham uma média de zero e um desvio padrão de um.

#### **Benefícios da Escala de Recursos**

1. **Melhoria na Convergência**: Com recursos dimensionados de forma semelhante, o gradiente descendente pode proceder de forma mais direta para o mínimo, melhorando a velocidade de convergência.
2. **Contornos mais Simétricos**: Os contornos da função de custo tornam-se mais simétricos e circulares, facilitando para o gradiente descendente avançar diretamente para o mínimo.

````python
# Função para calcular a função de custo
def compute_cost(X, y, w, b):
    m = len(y)
    predictions = np.dot(X, w) + b
    sqrErrors = (predictions - y) ** 2
    return 1 / (2 * m) * np.sum(sqrErrors)

# Gerar grades para w e b para plotar contornos da função de custo
def generate_cost_contours(X, y, w_range, b_range):
    W, B = np.meshgrid(w_range, b_range)
    Costs = np.zeros(W.shape)

    for i in range(W.shape[0]):
        for j in range(W.shape[1]):
            Costs[i, j] = compute_cost(X, y, W[i, j], B[i, j])
    
    return W, B, Costs

# Dados de exemplo
X = np.array([[300, 1500, 2000]])  # tamanho da casa
y = np.array([200, 400, 500])  # preço em milhares

# Normalização dos dados
X_norm = (X - np.min(X)) / (np.max(X) - np.min(X)) # normalização min-max

# Definindo intervalos para w e b
w_range = np.linspace(-10, 10, 100)
b_range = np.linspace(-100, 100, 100)

# Calculando contornos da função de custo
W, B, Costs_unscaled = generate_cost_contours(X, y, w_range, b_range)
_, _, Costs_scaled = generate_cost_contours(X_norm, y, w_range, b_range)

# Plotar contornos da função de custo
plt.figure(figsize=(12, 6))

# Para dados não normalizados
plt.subplot(1, 2, 1)
plt.contourf(W, B, Costs_unscaled, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Função de Custo Sem Normalização')
plt.xlabel('w')
plt.ylabel('b')

# Para dados normalizados
plt.subplot(1, 2, 2)
plt.contourf(W, B, Costs_scaled, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Função de Custo Com Normalização')
plt.xlabel('w')
plt.ylabel('b')

plt.show()
````

### Como Escolher uma Boa Taxa de Aprendizado

1. **Comece Pequeno**: Comece com uma taxa de aprendizado relativamente pequena (por exemplo, 0.001) para garantir que a função de custo diminua de maneira estável.

2. **Aumento Incremental**: Aumente gradualmente a taxa de aprendizado (por exemplo, 0.003, 0.01, 0.03, etc.) e observe o comportamento da função de custo. O objetivo é encontrar a maior taxa de aprendizado que ainda resulte em uma diminuição consistente na função de custo.

3. **Inspeção Visual**: Plote a função de custo contra o número de iterações para cada taxa de aprendizado. Essa inspeção visual pode rapidamente revelar se a taxa de aprendizado está muito alta (função de custo aumenta ou flutua significativamente) ou muito baixa (diminuição lenta).

4. **Depuração com um Alpha Pequeno**: Se a função de custo não estiver diminuindo consistentemente, reduza a taxa de aprendizado para um número muito pequeno como uma etapa de depuração. Se a função de custo ainda não diminuir, isso pode indicar um bug na implementação do gradiente descendente.

5. **Ajuste Baseado no Desempenho**: Dependendo dos recursos computacionais e do tempo disponível, ajuste a taxa de aprendizado para equilibrar a velocidade de convergência e a eficiência computacional.

````python
# Função para calcular a função de custo
def compute_cost(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    return (1 / (2 * m)) * np.sum((predictions - y) ** 2)

# Função de gradiente descendente
def gradient_descent(X, y, theta, alpha, num_iters):
    m = len(y)
    J_history = np.zeros(num_iters)

    for i in range(num_iters):
        predictions = X.dot(theta)
        errors = predictions - y
        updates = (alpha / m) * X.T.dot(errors)
        theta -= updates
        J_history[i] = compute_cost(X, y, theta)
    
    return theta, J_history

# Gerar dados sintéticos
np.random.seed(42)
X = np.random.rand(100, 3)
X[:, 0] *= 100   # Escala diferente para a primeira característica
y = X @ np.array([3, 200, 50]) + np.random.randn(100) * 50  # y é uma combinação linear de X com ruído

# Normalizando características
X_norm = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Adicionando coluna de 1s para intercepto
X = np.hstack([np.ones((X.shape[0], 1)), X])
X_norm = np.hstack([np.ones((X.shape[0], 1)), X_norm])

# Parâmetros iniciais
theta = np.zeros(X.shape[1])

# Taxas de aprendizado para teste
learning_rates = [0.001, 0.01, 0.05, 0.1]

# Plotando o impacto das taxas de aprendizado na convergência
plt.figure(figsize=(10, 6))

for alpha in learning_rates:
    _, J_history = gradient_descent(X_norm, y, theta, alpha, 100)
    plt.plot(J_history, label=f'alpha = {alpha}')

plt.xlabel('Número de Iterações')
plt.ylabel('Custo J')
plt.title('Efeito das Taxas de Aprendizado na Convergência')
plt.legend()
plt.show()
````

# Regressão Polinomial

A regressão polinomial é um tipo de análise de regressão onde a relação entre a variável independente $ x $ e a variável dependente $ y $ é modelada como um polinômio de grau $ n $. A regressão polinomial pode ajustar uma ampla gama de curvaturas em um conjunto de dados, tornando-a incrivelmente útil para modelar relações não lineares onde uma simples regressão linear não pode fornecer uma explicação satisfatória dos dados.

### Vantagens da Regressão Polinomial
1. **Flexibilidade**: Pode modelar dados com relações não lineares.
2. **Captura de Pontos de Infleção**: Pode considerar pontos no dados onde as tendências aumentam, diminuem ou se estabilizam.
3. **Melhor Ajuste**: Para muitos conjuntos de dados, os ajustes polinomiais podem fornecer uma aproximação muito melhor do que um ajuste linear simples.

### Implementando Regressão Polinomial
Para implementar a regressão polinomial, muitas vezes é necessário criar recursos que incluam potências dos recursos originais e possivelmente termos de interação. Veja como você pode configurá-lo para um conjunto de dados imobiliários onde o tamanho da casa influencia o preço de forma não linear:

````python
def create_poly_features(X, degree):
    """ Cria características polinomiais para uma única variável de entrada. """
    X_poly = np.ones((len(X), 1))  # Iniciar com a coluna de intercepto
    for i in range(1, degree + 1):
        X_poly = np.hstack((X_poly, X**i))
    return X_poly

def fit_polynomial_regression(X, y):
    """ Calcula os coeficientes usando a pseudo-inversa. """
    X_b = np.hstack([np.ones((len(X), 1)), X])  # Adiciona coluna para o intercepto
    w = np.linalg.pinv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
    return w

# Dados de exemplo
np.random.seed(0)
X = 2 * np.random.rand(100, 1) * 1000  # Tamanho da casa em pés quadrados
y = 3 * (X**2) - 0.1 * X**3 + np.random.randn(100, 1) * 15000  # Preços relacionados de forma não linear

# Preparar os dados para regressões polinomiais
X_poly_2 = create_poly_features(X, 2)
X_poly_3 = create_poly_features(X, 3)

# Ajustar os modelos
w_2 = fit_polynomial_regression(X_poly_2, y)
w_3 = fit_polynomial_regression(X_poly_3, y)

# Função para fazer previsões
def predict(X, w):
    X_b = np.hstack([np.ones((len(X), 1)), X])  # Adiciona coluna de intercepto
    return X_b.dot(w)

# Gerar dados para previsão
X_fit = np.linspace(0, 2000, 100).reshape(-1, 1)
X_fit_poly_2 = create_poly_features(X_fit, 2)
X_fit_poly_3 = create_poly_features(X_fit, 3)

y_pred_2 = predict(X_fit_poly_2, w_2)
y_pred_3 = predict(X_fit_poly_3, w_3)

# Plotar os resultados
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='gray', label='Dados reais')
plt.plot(X_fit, y_pred_2, label='Ajuste Quadrático', color='blue')
plt.plot(X_fit, y_pred_3, label='Ajuste Cúbico', color='green')
plt.title('Regressão Polinomial sem scikit-learn')
plt.xlabel('Tamanho da Casa (pés quadrados)')
plt.ylabel('Preço ($)')
plt.legend()
plt.show()
````


### Considerações Importantes
- **Escalonamento de Recursos**: Como destacado, ao usar termos polinomiais, especialmente de graus mais altos, o escalonamento de recursos se torna crucial para prevenir instabilidade numérica durante a otimização.
- **Risco de Overfitting**: Polinômios de grau mais alto podem se ajustar muito bem aos dados de treinamento, mas podem ter desempenho ruim em dados não vistos devido ao overfitting. É vital equilibrar a complexidade do modelo com a generalização.
- **Seleção de Modelo**: Usar validação cruzada para selecionar o grau do polinômio pode ajudar a evitar overfitting e underfitting.

# Classificação

Em problemas de classificação, a variável de saída $ y $ pode assumir apenas um número limitado de valores. Ao contrário da regressão linear, onde $ y $ pode assumir qualquer valor num intervalo contínuo, a classificação lida com categorias discretas. Por exemplo, classificar se um email é spam (1) ou não (0).

### Inadequação da Regressão Linear para Classificação

A regressão linear não é adequada para classificação principalmente porque pode prever valores fora do intervalo necessário (por exemplo, menor que 0 ou maior que 1 para uma classificação binária). Além disso, a regressão linear é sensível a outliers, o que pode distorcer significativamente a linha de melhor ajuste, como demonstrado no exemplo do tumor.

### Problema com Regressão Linear: Um Exemplo com Tumor

Suponhamos que temos um conjunto de dados onde $ y $ representa se um tumor é maligno (1) ou benigno (0), e $ x $ é o tamanho do tumor. Usando a regressão linear, poderíamos tentar ajustar uma linha para prever $ y $ com base em $ x $. No entanto, isso pode levar a problemas significativos:

1. **Sensibilidade a Outliers**: Um único ponto de dados distante pode alterar drasticamente a linha de regressão, como um tumor muito grande que ainda é benigno.
2. **Valores Inadequados**: A linha de regressão linear pode prever valores fora de 0 e 1, o que não faz sentido em um contexto de classificação.

### Introdução à Regressão Logística

Para superar essas limitações, introduzimos a **regressão logística**, que é adequada para problemas de classificação. A regressão logística usa a função sigmoide para garantir que a saída esteja sempre entre 0 e 1, tornando-a ideal para prever probabilidades.

### Funcionamento da Regressão Logística

A função sigmoide é expressa como:
$$ \sigma(z) = \frac{1}{1 + e^{-z}} $$
onde $ z = w \cdot x + b $.

Na regressão logística, usamos um **limiar** (geralmente 0,5) para decidir a classe de $ y $:
- Se $ \sigma(z) \geq 0.5 $, $ y $ é classificado como 1.
- Se $ \sigma(z) < 0.5 $, $ y $ é classificado como 0.



````python
# Exemplo de dados de tumores
x = np.array([1, 2, 3, 20])  # Tamanho do tumor
y = np.array([0, 0, 0, 1])   # 0 = benigno, 1 = maligno

# Ajustando a regressão linear
w_linear = np.polyfit(x, y, 1)
y_linear = np.poly1d(w_linear)(x)

# Ajustando a regressão logística
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

w_log = np.array([0.1, -1.8])  # Coeficientes inventados para exemplo
z_log = w_log[0] * x + w_log[1]
y_log = sigmoid(z_log)

# Plotando
plt.scatter(x, y, color='red', label='Dados Reais')
plt.plot(x, y_linear, label='Regressão Linear', color='blue')
plt.plot(x, y_log, label='Regressão Logística', color='green')
plt.xlabel('Tamanho do Tumor')
plt.ylabel('Probabilidade de Malignidade')
plt.legend()
plt.show()
````

### Problemas de Classificação

Aqui estão alguns exemplos de problemas de classificação: tentar descobrir se um e-mail é spam, então a resposta que você quer é um sim ou não. Outro exemplo seria descobrir se uma transação financeira online é fraudulenta. Combater fraudes financeiras online é algo em que já trabalhei, e era estranhamente emocionante porque eu sabia que havia forças tentando roubar dinheiro e o trabalho da minha equipe era impedi-las.

O problema é dado uma transação financeira. Seu algoritmo de aprendizado pode descobrir se esta transação é fraudulenta, como se esse cartão de crédito fosse roubado? Outro exemplo que tocamos antes foi tentar classificar um tumor como maligno versus benigno. Em cada um desses problemas, a variável que você deseja prever pode ser apenas um de dois valores possíveis: Sim ou Não. Esse tipo de problema de classificação onde existem apenas dois resultados possíveis é chamado de classificação binária, onde a palavra binária refere-se a haver apenas duas classes ou categorias possíveis.

### Regressão Logística

Em contraste com a regressão linear, na regressão logística, o que fazemos é ajustar uma curva em forma de S aos dados, como mostrado na imagem a seguir. Por exemplo, se um paciente chega com um tumor do tamanho mostrado no eixo x, então o algoritmo produzirá 0,7, sugerindo que é mais provável que seja maligno do que benigno. Discutiremos mais tarde o que exatamente significa 0,7 neste contexto, mas é importante notar que a saída do modelo nunca é realmente 0,7; é sempre 0 ou 1.

Para construir o algoritmo de regressão logística, existe uma função matemática importante que gostaria de descrever, chamada função Sigmóide, às vezes também referida como função logística. A função sigmóide se parece com isto:

![Sigmoid Function](https://i.imgur.com/g7lA21h.png)

Se usarmos $ g(z) $ para denotar essa função, a fórmula de $ g(z) $ é $ \frac{1}{1 + e^{-z}} $, onde $ e $ é uma constante matemática que vale aproximadamente 2,7. Se $ z $ for muito grande, $ e^{-z} $ será um número muito pequeno, então $ g(z) $ será muito próximo de 1. Por outro lado, se $ z $ for um número negativo muito grande, então $ g(z) $ será muito próximo de 0.

### Implementação da Regressão Logística

O modelo de regressão logística é definido como $ f(x) = g(w \cdot x + b) $, onde $ w $ são os pesos, $ x $ são as características, $ b $ é o viés e $ g $ é a função sigmóide. Essa saída é interpretada como a probabilidade de $ y = 1 $ dado $ x $ com parâmetros $ w $ e $ b $.

Para entender como a regressão logística faz previsões, vamos visualizar o conceito de "fronteira de decisão". Vamos considerar um problema de classificação com duas características, $ x_1 $ e $ x_2 $, e definir $ z = w_1x_1 + w_2x_2 + b $. A fronteira de decisão ocorre onde $ z = 0 $. Se $ w_1 = 1 $, $ w_2 = 1 $, e $ b = -3 $, a linha de decisão seria $ x_1 + x_2 = 3 $.

````python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def compute_cost(X, y, w, b):
    m = len(y)
    z = np.dot(X, w) + b
    predictions = sigmoid(z)
    # Função de custo para a regressão logística binária (entropia cruzada)
    cost = -np.sum(y * np.log(predictions) + (1 - y) * np.log(1 - predictions)) / m
    return cost

def gradient_descent(X, y, w, b, alpha, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        z = np.dot(X, w) + b
        predictions = sigmoid(z)
        error = predictions - y
        w -= (alpha / m) * np.dot(X.T, error)
        b -= (alpha / m) * np.sum(error)
        cost_history.append(compute_cost(X, y, w, b))

    return w, b, cost_history

# Gerar dados sintéticos
np.random.seed(0)
x1 = 2 * np.random.rand(100, 1)
x2 = 2 * np.random.rand(100, 1)
y = (x1**2 + x2**2 <= 1).astype(int).flatten()

# Preparar dados para regressão logística
X = np.hstack((x1, x2))
initial_w = np.zeros(X.shape[1])
initial_b = 0

# Configurações para o gradiente descendente
alpha = 1
iterations = 1000

# Executando o gradiente descendente
final_w, final_b, cost_history = gradient_descent(X, y, initial_w, initial_b, alpha, iterations)

# Plotar o resultado
plt.figure(figsize=(10, 6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', label='Classe 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Classe 1')

# Definir fronteira de decisão
x_values = [np.min(X[:, 0] - 0.5), np.max(X[:, 0] + 0.5)] 
y_values = -(final_b + np.dot(final_w[0], x_values)) / final_w[1]
plt.plot(x_values, y_values, label='Fronteira de Decisão')

plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.title('Regressão Logística com Fronteira de Decisão')
plt.legend()
plt.show()
````

Na regressão logística, diferentemente da regressão linear, a função de custo de erro quadrático não é apropriada devido à sua natureza não convexa quando aplicada à saída da função sigmóide. Isso pode levar a muitos mínimos locais, complicando o uso do gradiente descendente para encontrar os parâmetros ótimos.

### Explicação Detalhada:

1. **Conjunto de Treinamento e Modelo**:
   - Em um problema de classificação binária como a regressão logística, cada exemplo de treinamento tem um ou mais atributos (como tamanho do tumor, idade do paciente, etc.) e um rótulo de destino que pode ser 0 ou 1.
   - O modelo de regressão logística é dado pela equação $ f(x) = \frac{1}{1 + e^{-(w \cdot x + b)}} $, onde $ w $ e $ b $ são os parâmetros do modelo.

2. **Função de Custo na Regressão Linear**:
   - Na regressão linear, a função de custo de erro quadrático é dada por $ J(w, b) = \frac{1}{2m} \sum_{i=1}^m (f(x^{(i)}) - y^{(i)})^2 $, onde $ f(x) = w \cdot x + b $.
   - Essa função de custo é convexa, garantindo que o gradiente descendente converge para o mínimo global.

3. **Problema com Erro Quadrático na Regressão Logística**:
   - Se tentarmos usar a mesma função de custo para a regressão logística substituindo $ f(x) $ pela função sigmóide, a função de custo se torna não convexa. Isso significa que existem múltiplos mínimos locais, e o gradiente descendente pode não encontrar o mínimo global, ficando preso em um mínimo local.

4. **Nova Função de Custo para Regressão Logística**:
   - Para superar o problema da não convexidade, utilizamos uma função de custo diferente na regressão logística, que é convexa e permite que o gradiente descendente funcione eficazmente.
   - A função de custo, chamada de função de custo logística ou entropia cruzada, é definida como:
     - $ J(w, b) = -\frac{1}{m} \sum_{i=1}^m [y^{(i)} \log(f(x^{(i)})) + (1 - y^{(i)}) \log(1 - f(x^{(i)}))] $
   - Essa função penaliza fortemente as classificações erradas, especialmente aquelas que estão confiantes e erradas.

5. **Detalhes da Função de Custo Logística**:
   - Se o rótulo verdadeiro $ y $ é 1, a perda é $ -\log(f(x)) $, que se torna grande se $ f(x) $ (probabilidade prevista de 1) for pequena.
   - Se $ y $ é 0, a perda é $ -\log(1 - f(x)) $, que se torna grande se $ f(x) $ for grande.

````python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Gerar dados de exemplo
x = np.linspace(-10, 10, 100)
y = np.where(x >= 0, 1, 0)  # Rótulos: 1 se x >= 0, senão 0
w_vals = np.linspace(-10, 10, 100)
b_vals = np.linspace(-10, 10, 100)
W, B = np.meshgrid(w_vals, b_vals)

# Calcular previsões e custos para cada par (w, b)
SE_costs = np.zeros(W.shape)
CE_costs = np.zeros(W.shape)

for i in range(W.shape[0]):
    for j in range(W.shape[1]):
        w = W[i, j]
        b = B[i, j]
        z = w * x + b
        predictions = sigmoid(z)
        # Custo de erro quadrático
        SE_costs[i, j] = np.mean(0.5 * (predictions - y) ** 2)
        # Evitar log(0) nas previsões
        predictions = np.clip(predictions, 1e-15, 1 - 1e-15)
        # Custo de entropia cruzada
        CE_costs[i, j] = -np.mean(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))

# Plotar
fig = plt.figure(figsize=(14, 6))

ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(W, B, SE_costs, cmap='viridis')
ax.set_title('Custo de Erro Quadrático')
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('Custo')

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(W, B, CE_costs, cmap='viridis')
ax2.set_title('Custo de Entropia Cruzada')
ax2.set_xlabel('w')
ax2.set_ylabel('b')
ax2.set_zlabel('Custo')

plt.show()
````

Vamos agora explorar uma maneira simplificada de expressar a função de perda e a função de custo na regressão logística, facilitando a implementação quando chegarmos ao gradiente descendente para ajustar os parâmetros do modelo. Como um lembrete, aqui está a função de perda que definimos anteriormente para a regressão logística. Dado que estamos lidando com um problema de classificação binária, `y` pode assumir apenas os valores zero ou um.

Podemos reescrever a função de perda de forma mais simplificada da seguinte maneira:
$$ \text{Perda} = -y \log(f(x)) - (1 - y) \log(1 - f(x)) $$

Essa equação, que escrevemos em uma única linha, é completamente equivalente à fórmula mais complexa mencionada anteriormente. Vejamos por que isso é verdadeiro, lembrando que `y` só pode ser 0 ou 1:

- **Caso `y = 1`**:
  - O termo `y` é 1 e o termo `(1 - y)` é 0.
  - Assim, a perda se torna $-1 \times \log(f(x)) - 0 \times \text{alguma coisa}$, que simplifica para $-\log(f(x))$.

- **Caso `y = 0`**:
  - O termo `y` é 0, então o primeiro termo se anula.
  - A perda se torna $-1 \times \log(1 - f(x))$, que é igual ao segundo termo da expressão original.

Isso demonstra que, independentemente do valor de `y`, essa única expressão cobre ambos os casos, simplificando a função de perda sem necessidade de separar os dois casos.

Com essa função de perda simplificada, vamos revisitar a função de custo para a regressão logística. A função de custo `J` é simplesmente a média das perdas em todo o conjunto de treinamento de `m` exemplos:
$$ J = \frac{1}{m} \sum_{i=1}^m \text{Perda}_i $$

Se inserirmos a definição da função de perda simplificada acima, a expressão fica:
$$ J = \frac{1}{m} \sum_{i=1}^m \left(-y_i \log(f(x_i)) - (1 - y_i) \log(1 - f(x_i))\right) $$

Ao mover os sinais negativos para fora do somatório, obtemos:
$$ J = -\frac{1}{m} \sum_{i=1}^m \left(y_i \log(f(x_i)) + (1 - y_i) \log(1 - f(x_i))\right) $$

Essa função de custo, amplamente usada para treinar regressão logística, é derivada de um princípio estatístico chamado estimativa de máxima verossimilhança. Este princípio é uma ideia da estatística sobre como encontrar eficientemente parâmetros para diferentes modelos. Além disso, esta função de custo possui a propriedade desejável de ser convexa, o que é crucial para a eficácia do gradiente descendente.

Para ajustar os parâmetros do modelo de regressão logística, precisamos encontrar os valores dos parâmetros $ w $ e $ b $ que minimizem a função de custo $ J(w, b) $. Para isso, utilizaremos o método do gradiente descendente. Neste vídeo, vamos focar em como encontrar uma boa escolha para os parâmetros $ w $ e $ b $. Depois de definidos, se fornecermos ao modelo um novo input $ x $, como por exemplo os dados de um paciente com um certo tamanho de tumor e idade, o modelo poderá fazer uma previsão ou tentar estimar a probabilidade de que o label $ y $ seja 1.

O algoritmo de gradiente descendente é usado para minimizar a função de custo $ J $. Aqui está a função de custo novamente. Para minimizar $ J $ como uma função de $ w $ e $ b $, utilizamos a seguinte regra de atualização no gradiente descendente, onde cada parâmetro é atualizado pelo valor anterior menos o produto da taxa de aprendizado $ \alpha $ pelo termo derivativo:

$$
w_j := w_j - \alpha \left(\frac{1}{m} \sum_{i=1}^m (f(x^{(i)}) - y^{(i)}) x_j^{(i)}\right)
$$

$$
b := b - \alpha \left(\frac{1}{m} \sum_{i=1}^m (f(x^{(i)}) - y^{(i)})\right)
$$

Aqui, $ x_j^{(i)} $ é a $ j $-ésima característica do $ i $-ésimo exemplo de treinamento. A função $ f(x) $ na regressão logística é definida como a função sigmóide aplicada a $ wx + b $, diferentemente da regressão linear onde $ f(x) $ é simplesmente $ wx + b $.

Essas derivadas são essencialmente as mesmas que as usadas na regressão linear, mas é crucial notar que mesmo que as expressões para as atualizações dos parâmetros se pareçam, os algoritmos são distintos porque a função $ f(x) $ na regressão logística passa pelo filtro da função sigmóide, o que não ocorre na regressão linear.

**Underfitting**, ou subajuste, ocorre quando o modelo é muito simples para capturar a complexidade dos dados. No exemplo dado, um modelo linear (uma linha reta) pode não se ajustar bem ao conjunto de dados porque, conforme o tamanho da casa aumenta, os preços tendem a se estabilizar. Esse modelo linear tem **alta viés** (high bias), significando que ele não se ajusta suficientemente bem nem mesmo aos dados de treinamento, falhando em capturar a tendência geral.

A **alta viés** é uma indicação de que o modelo tem preconcepções fortes sobre a forma dos dados (neste caso, que a relação é linear) que não correspondem à realidade, levando a erros sistemáticos tanto nos dados de treinamento quanto nos novos dados, ou seja, o modelo tem baixa capacidade de generalização.

Por outro lado, **overfitting**, ou sobreajuste, ocorre quando um modelo é excessivamente complexo, contendo mais parâmetros do que o necessário para descrever os dados. Isso pode ser visto quando ajustamos um polinômio de quarta ordem aos dados de preços das casas, permitindo que o modelo se ajuste perfeitamente aos pontos de dados de treinamento. Embora isso possa parecer bom inicialmente, o modelo se torna tão específico para os dados de treinamento que é incapaz de generalizar bem para novos exemplos. Esse modelo tem **alta variância**, indicando que pequenas variações nos dados de treinamento podem resultar em grandes variações no comportamento do modelo.

Modelos com **alta variância** são sensíveis ao ruído nos dados de treinamento, e pequenas mudanças nos dados podem levar a mudanças significativas no modelo, o que pode não ser desejável.

Aqui estão três estratégias principais para lidar com o overfitting:

1. **Coletar mais dados**: Aumentar o conjunto de dados de treinamento pode ajudar o modelo a aprender padrões mais gerais e robustos. Quando você tem mais dados, o modelo tem a oportunidade de aprender a distinguir entre variações de dados genuínas e ruído aleatório, o que pode reduzir o overfitting.

2. **Seleção de características (Feature Selection)**: Reduzir o número de características usadas no modelo pode ajudar a prevenir overfitting, especialmente se algumas características forem irrelevantes ou redundantes. Ao concentrar-se nas características mais informativas, o modelo pode aprender padrões mais significativos sem ser distraído por ruído.

3. **Regularização**: Essa técnica modifica a função de custo que o modelo minimiza, adicionando um termo de penalidade aos parâmetros do modelo. Isso desencoraja o modelo de aprender parâmetros extremamente grandes, o que geralmente está associado a curvas complexas que se ajustam demais aos dados. A regularização suaviza o modelo, incentivando-o a aprender uma função mais simples.

   - A regularização pode ser implementada de várias formas, como L1 (Lasso), que penaliza a soma dos valores absolutos dos coeficientes, e L2 (Ridge), que penaliza a soma dos quadrados dos coeficientes.

# Regularização

**Função de Custo Modificada**: Ao invés de apenas minimizar o erro (por exemplo, o erro quadrático médio em regressão linear), adicionamos um termo de regularização que penaliza a magnitude dos coeficientes do modelo.

**Termo de Regularização**: O termo de regularização mais comum é o termo L2, que é a soma dos quadrados dos coeficientes, multiplicada por um parâmetro de regularização λ (lambda). A função de custo modificada fica então:
   
   $$
   J(w) = \frac{1}{2m} \sum_{i=1}^{m} (f_w(x^{(i)}) - y^{(i)})^2 + \frac{\lambda}{2m} \sum_{j=1}^{n} w_j^2
   $$

   Aqui, $ m $ é o número de exemplos de treinamento, $ n $ é o número de características, $ w $ são os parâmetros do modelo, $ f_w(x^{(i)}) $ é a predição do modelo para o i-ésimo exemplo de treinamento, e $ y^{(i)} $ é o verdadeiro valor de saída para o i-ésimo exemplo de treinamento.

**Impacto da Regularização**:
   - **λ grande**: Um valor grande para λ penaliza fortemente os coeficientes, podendo levar a um modelo muito simples que pode subajustar os dados (underfitting).
   - **λ pequeno**: Um valor pequeno para λ coloca menos ênfase na penalização dos coeficientes, podendo levar a um modelo mais complexo que pode sobreajustar os dados (overfitting).

**Escolhendo λ**: O valor de λ é um hiperparâmetro que você precisa escolher usando técnicas como validação cruzada. Não há um valor único de λ que funcione melhor para todos os problemas, então a escolha geralmente envolve experimentação e ajuste baseado no desempenho do modelo em um conjunto de validação.

**Efeito no Treinamento**: A inclusão do termo de regularização altera as derivadas da função de custo, que são usadas no algoritmo de descida do gradiente para atualizar os parâmetros. As atualizações dos parâmetros levam em conta não apenas o erro em relação à previsão, mas também o tamanho dos próprios parâmetros.

````python
# Gerar dados sintéticos
np.random.seed(0)
x = np.linspace(0, 1, 5)  # Poucos pontos para aumentar chance de overfitting
y = 1 + x**2 + np.random.normal(0, 0.2, size=x.shape)

# Função para criar características polinomiais
def create_poly_features(x, degree):
    return np.array([x**i for i in range(degree + 1)]).T

# Gradiente descendente com regularização L2
def gradient_descent(X, y, alpha, epochs, lambda_):
    m, n = X.shape
    w = np.zeros(n)  # Inicializar parâmetros
    J_history = []  # Armazenar o custo em cada etapa
    
    for epoch in range(epochs):
        predictions = X.dot(w)
        errors = predictions - y
        gradient = (X.T.dot(errors) + lambda_ * w) / m
        w -= alpha * gradient
        
        # Calcular a função de custo com termo de regularização
        cost = (np.sum(errors**2) + lambda_ * np.sum(w**2)) / (2 * m)
        J_history.append(cost)
    
    return w, J_history

# Configurações do modelo
alpha = 0.1  # Taxa de aprendizado
epochs = 100000  # Número de iterações reduzidas para prática
degree = 20  # Grau do polinômio

# Treinar modelos com diferentes valores de lambda
lambdas = [0, 0.1, 0.01]
models = {}
for lambda_ in lambdas:
    X_poly = create_poly_features(x, degree)
    models[lambda_], _ = gradient_descent(X_poly, y, alpha, epochs, lambda_)

# Plotar os dados e os modelos
plt.figure(figsize=(12, 8))
plt.scatter(x, y, color='gray', label='Dados')
x_range = np.linspace(0, 1, 100)

for lambda_, w in models.items():
    y_pred = np.sum([w[i] * x_range**i for i in range(degree + 1)], axis=0)
    plt.plot(x_range, y_pred, label=f'Lambda = {lambda_}')

plt.title('Comparação de Modelos com Diferentes Valores de Lambda')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
````

Para implementar o gradiente descendente em regressão linear regularizada, precisamos ajustar nossa função de custo e o processo de atualização dos parâmetros para incluir o termo de regularização. Esse termo ajuda a prevenir o sobreajuste (overfitting) reduzindo a magnitude dos coeficientes da regressão. Vamos ver como fazer isso.

### Função de Custo para Regressão Linear Regularizada

A função de custo regularizada $ J(w, b) $ para a regressão linear é definida como:

$$
J(w, b) = \frac{1}{2m} \left[ \sum_{i=1}^m (w \cdot x^{(i)} + b - y^{(i)})^2 + \lambda \sum_{j=1}^n w_j^2 \right]
$$

onde:
- $ m $ é o número de exemplos de treinamento,
- $ x^{(i)} $ são os atributos do i-ésimo exemplo de treinamento,
- $ y^{(i)} $ é o rótulo do i-ésimo exemplo de treinamento,
- $ w $ e $ b $ são os parâmetros do modelo,
- $ \lambda $ é o parâmetro de regularização,
- $ n $ é o número de atributos.

### Gradiente Descendente para Regressão Linear Regularizada

As atualizações para os parâmetros são feitas conforme as seguintes equações:

- Para $ w_j $ (cada coeficiente do modelo exceto o intercepto):
  $$
  w_j := w_j - \alpha \left( \frac{1}{m} \sum_{i=1}^m ((w \cdot x^{(i)} + b - y^{(i)}) \cdot x_j^{(i)} + \lambda w_j) \right)
  $$
  onde $ x_j^{(i)} $ é o valor do j-ésimo atributo do i-ésimo exemplo de treinamento.

- Para $ b $ (o intercepto):
  $$
  b := b - \alpha \left( \frac{1}{m} \sum_{i=1}^m (w \cdot x^{(i)} + b - y^{(i)}) \right)
  $$

````python
def gradient_descent_regularized(X, y, w, b, alpha, num_iters, lambda_):
    m = len(y)
    for i in range(num_iters):
        predictions = X.dot(w) + b
        errors = predictions - y
        gradient_w = (X.T.dot(errors) + lambda_ * w) / m
        gradient_b = np.sum(errors) / m
        w -= alpha * gradient_w
        b -= alpha * gradient_b
    return w, b

X = np.array([[1, 2], [1, 3], [1, 4], [1, 5]])
y = np.array([7, 6, 5, 4])
w = np.array([0.1, 0.2])
b = 0.3
alpha = 0.1
num_iters = 1000
lambda_ = 0.1
w, b = gradient_descent_regularized(X, y, w, b, alpha, num_iters, lambda_)

# Plot
plt.scatter(X[:, 1], y, color='blue', label='Dados')
plt.plot(X[:, 1], X.dot(w) + b, color='red', label='Linha de Regressão')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
````

### Função de Custo da Regressão Logística Regularizada

A função de custo para regressão logística regularizada inclui o termo de regularização L2 (norma quadrada dos pesos), que penaliza valores grandes dos coeficientes $w$. A função de custo $J(w, b)$ é definida como:

$$
J(w, b) = -\frac{1}{m} \sum_{i=1}^m \left[ y^{(i)} \log(f(x^{(i)})) + (1 - y^{(i)}) \log(1 - f(x^{(i)})) \right] + \frac{\lambda}{2m} \sum_{j=1}^n w_j^2
$$

onde:
- $m$ é o número de exemplos de treinamento.
- $y^{(i)}$ é o rótulo da i-ésima observação.
- $f(x^{(i)}) = \sigma(w \cdot x^{(i)} + b)$ é a previsão do modelo, com $\sigma(z)$ sendo a função sigmoide.
- $\lambda$ é o parâmetro de regularização.
- $n$ é o número de parâmetros (excluindo o intercepto).

### Atualização dos Parâmetros via Gradiente Descendente

No gradiente descendente para a regressão logística regularizada, as atualizações dos parâmetros são dadas por:

- Para os pesos $w_j$ (para cada característica $j$):
  $$
  w_j := w_j - \alpha \left(\frac{1}{m} \sum_{i=1}^m (f(x^{(i)}) - y^{(i)}) x_j^{(i)} + \frac{\lambda}{m} w_j\right)
  $$

- Para o intercepto $b$:
  $$
  b := b - \alpha \left(\frac{1}{m} \sum_{i=1}^m (f(x^{(i)}) - y^{(i)})\right)
  $$

````python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def compute_cost(X, y, w, b, lambda_):
    m = len(y)
    z = X.dot(w) + b
    predictions = sigmoid(z)
    cost = (-y * np.log(predictions) - (1 - y) * np.log(1 - predictions)).mean()
    reg_cost = (lambda_ / (2 * m)) * np.sum(w**2)
    return cost + reg_cost

def gradient_descent(X, y, w, b, alpha, num_iters, lambda_):
    m = len(y)
    for i in range(num_iters):
        z = X.dot(w) + b
        predictions = sigmoid(z)
        errors = predictions - y
        w_gradient = (X.T.dot(errors) + lambda_ * w) / m
        b_gradient = np.sum(errors) / m
        w -= alpha * w_gradient
        b -= alpha * b_gradient
    return w, b

# Exemplo de uso
X = np.array([[1, 2], [1, 3], [1, 4]])  # Matriz de características
y = np.array([0, 1, 1])                # Vetor de rótulos
w = np.zeros(X.shape[1])               # Inicialização dos pesos
b = 0                                  # Inicialização do intercepto
alpha = 0.1                            # Taxa de aprendizado
num_iters = 1000                       # Número de iterações
lambda_ = 0.01                         # Parâmetro de regularização

w, b = gradient_descent(X, y, w, b, alpha, num_iters, lambda_)

# Plot
plt.scatter(X[:, 1], y, color='blue', label='Dados')
plt.plot(X[:, 1], sigmoid(X.dot(w) + b), color='red', label='Linha de Regressão')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
````
